# Q1 logisch gevolg

* We werken met de [sympy module](https://docs.sympy.org/latest/modules/logic.html). Dat is erg handig. Neem hem eerst goed door en speel er wat mee.
* In de cel hieronder vind je een handige routine.
* Je gebruikt natuurlijk  de definitie van $\Sigma\models\phi$, van logisch gevolg dus.
* En daarvoor is de `sympy` routine `satisfiable` best handig....
* De tests in de antwoord cel gaven bij mij dit:
<pre>
{p, q | ~p} q True
{q, q | ~p} p [{q: True, p: False}]
{q, Implies(p, q)} p {q: True, p: False}
{q & (r | s), Implies(p, q)} p [{q: True, r: True, s: True, p: False}, {q: True, r: True, s: False, p: False}, {q: True, s: True, r: False, p: False}]
WARNING: All proposition letters in the Conclusion should occur in some formula of the Premisses.
</pre>

```python
from sympy import *
from sympy.logic.inference import satisfiable

p,q,r,s,t = symbols('p q r s t')

def maakconjunctie(P):
    '''Creates a conjunction from a set or list of formulas.'''
    P=list(P)
    if P:
        return P[0]&maakconjunctie(P[1:])
    else:
        return True
#test
maakconjunctie({p, ~p |q, p|(p&q)})    
```

```python
def implies(Premisses, Conclusion, allmodels=False):
    '''Return True if Conclusion follows from Premisses, 
    return a (or all) countermodel(s) otherwise. '''
    P= maakconjunctie(Premisses)
    
    if not Conclusion.atoms().issubset(P.atoms()):
        print('WARNING: All proposition letters in the Conclusion should occur in some formula of the Premisses.')
    else:
        # Conclusion follows from Premisses if satisfiable(Conclusion) is a subset of all satisfiable(Premisses) models
        # --> counter_examples are when this is not the case
        sat_C = satisfiable(Conclusion).items()
        counter_examples = [model for model in satisfiable(P, all_models=True) if not(sat_C <= model.items())]
        return True if not counter_examples else counter_examples if allmodels and counter_examples else counter_examples[0]
     
     
#TEST
P= {~p|q,p}
C= q   

print(P, C, implies(P,C, allmodels=True ))
P= {~p|q,q}
C= p   

print(P, C, implies(P,C, allmodels=True ))

P= {p>>q,q}  # andere manier om implicatie te schrijven
C= p   

print(P, C, implies(P,C))

P= {p>>q,q,(r|s)}  # andere manier om implicatie te schrijven
C= p 

print(P, C, implies(P,C, allmodels=True ))
P= {p>>q,q}  # andere manier om implicatie te schrijven
C= r

implies(P,C) #Dit geeft een error. Probeer het ook zelf
```

```python
assert_equal(type(implies({p},p)), bool)
assert_equal(type(implies({p|q},p)), dict)
assert_equal(type(implies({p|q},p, allmodels=True)), list)
```

```python
# test graph G: undirected graph zonder attributen 
G= nx.karate_club_graph()
nx.draw_networkx(G)

# test query A: Geef me alle driehoekjes in G. Ik vond 270 verschillende.
A = nx.Graph()
A.add_edges_from([('x','y'),('x','z'),('y','z')])
```

```python
# Test graph H: directed graph met attributen
H= nx.DiGraph()

Relaties = [('Liesje', 'Tom'), ('Tom', 'Teun'), ('Teun', 'Liesje'),('Lotje', 'Tim'), ('Tim', 'Tom'), ('Tom', 'Lotje'), ('Lola', 'Lotje'), ('Lola', 'Liesje'), ('Liesje', 'Lotje') ]

H.add_edges_from(Relaties)

Mannen,Vrouwen= {'Tom', 'Tim', 'Teun'},{'Liesje', 'Lotje', 'Lola'}

for n in Mannen:
    H.nodes[n]['M']=True
    H.nodes[n]['V']=False
for n in Vrouwen:
    H.nodes[n]['V']=True 
    H.nodes[n]['M']=False
nx.draw_kamada_kawai(H, with_labels=True, node_color='#FFFF')



# test query B: directed graph met attributen
B = nx.DiGraph()
B.add_edges_from([('x','y'),('y','z'),('z','x')])
B.nodes['x']['V'] = True
B.nodes['x']['M'] = False
B.nodes['y']['M'] = True
B.nodes['y']['V'] = False
B.nodes['z']['M'] = True
B.nodes['z']['V'] = False

nx.draw_kamada_kawai(B, with_labels=True, node_color='#FFFF')
```

```python
from itertools import combinations


def CQ(Model,Query):
    '''Returns a generator of dicts, where each dict is a bijective mapping from nodes in Model
    to all nodes in  Query.
    The keys are the model nodes.
    The set of dict values equals the set of nodes of Query.
    The generator returns all answers to the Query on this Model.'''

    # verkijgen van combinaties van len(A) aantal nodes
    for nodes in combinations(Model.nodes, len(Query.nodes)):
        
        # isomorphe subgraphs maken van node combinatie
        GM = nx.isomorphism.GraphMatcher(Model.subgraph(nodes),Query)

        # lopen door isomorphe subgraphs: mapping heeft de vorm van {'Liesje': 'x', 'Tom': 'y', 'Teun': 'z'}
        for mapping in GM.subgraph_isomorphisms_iter():

            # lopen door mapping om te zorgen dat de attributen van de pairs gelijk zijn aan elkaar
            if all([Query.nodes[mapping[pair]].items() <= Model.nodes[pair].items() for pair in mapping.keys()]):
                yield mapping


#test
print(len(list(CQ(G,A))))
for antwoord in CQ(G,A):
    print(antwoord)
print(len(list(CQ(H,B))))
for antwoord in CQ(H,B):
    print(antwoord)
```
