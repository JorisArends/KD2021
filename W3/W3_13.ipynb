```python
from sympy import *
from sympy.logic.inference import satisfiable

p,q,r,s,t = symbols('p q r s t')

def maakconjunctie(P):
    '''Creates a conjunction from a set or list of formulas.'''
    P=list(P)
    if P:
        return P[0]&maakconjunctie(P[1:])
    else:
        return True
#test
maakconjunctie({p, ~p |q, p|(p&q)})    
```

Met de volgende logische zin verbieden we dus het feit dat een opa of oma niet tegelijkertijd een vader of moeder van hetzelfde kind kan zijn:


* $\forall xy( Kind (x,y) \leftrightarrow R(y,x) \wedge \exists z(R(y, z) \wedge \lnot (y = z)))$


- Volgens deze formule kan jouw opa of oma niet tegelijkertijd jouw vader of moeder zijn. Hiermee verbieden we dus het eerste model van de meegegeven afbeelding.

- In de afbeelding hierboven wordt weergegeven dat een kind twee mannelijke ouders / twee vaders kan hebben. Biologisch gezien kan dit niet, omdat je altijd een vrouw nodig hebt om een kind te krijgen. In de werkelijkheid kan je dus niet twee (biologische) vaders hebben. Deze zin verbieden wij als volgt:


* $\forall xy( Kind (x,y) \leftrightarrow ((R(y,x) \wedge V(x)) \wedge (\exists z(R(z,x) \wedge M(z)))) \vee ((R(y,x) \wedge M(y))\wedge (\exists z (R(z,x) \wedge V(z)))))$


- Volgens deze formule moet je te allen tijde een vader en een moeder hebben. Hiermee verbieden we dus het tweede model van de meegegeven afbeelding.

- In de afbeelding hierboven wordt aangegeven dat een kind ouder van zichzelf kan zijn. In de werkelijkheid kan dit natuurlijk helemaal niet. Deze zin verbieden wij als volgt:


* $\forall xy( Kind(x,y) \leftrightarrow R(y,x) \wedge \lnot (x = y) \wedge \exists z(R(z,x) \wedge \lnot (x = z)))$


- Volgens deze formule moet je te allen tijde twee ouders hebben, en kan jij dus niet een van je ouders zijn. Hiermee verbieden we dus het derde model van de meegegeven afbeelding.

- In de afbeelding hierboven wordt aangegeven dat een kind ouder kan zijn van zijn opa of oma. In de werkelijkheid kan dit natuurlijk helemaal niet, omdat je kind bent van je ouders, en kleinkind van je opa en oma. Dan kun je dus niet de ouder zijn van je opa of oma. Deze zin verbieden wij als volgt:


* $\forall xy( Kind (x,y) \leftrightarrow R(y,x) \wedge \exists vz(R(z,y) \wedge R(v,z) \wedge \lnot (x = v)))$


- Volgens deze formule heeft je opa of oma een vader of moeder, en kan jij die vader of moeder van je opa of oma niet zijn. Hiermee verbieden we dus het vierde model van de meegegeven afbeelding.

- In de afbeelding hierboven wordt aangegeven dat een kind ouder kan zijn van zijn of haar vader of moeder. In de werkelijkheid kan dit natuurlijk helemaal niet, omdat je kind bent van je ouders en dan dus geen ouder kan zijn van je vader of moeder. Deze zin verbieden wij als volgt:


* $\forall xy( Kind (x,y) \leftrightarrow R(y,x) \wedge \exists z(R(z,y) \wedge \lnot (x = z)))$ 


- Volgens deze formule ben je kind van je ouders. Je ouders hebben een vader of moeder, en jij bent natuurlijk niet de ouder van je vader of moeder (in de formule: x is niet gelijk aan z). Hiermee verbieden we dus het vijfde en laatste model van de meegegeven afbeelding.

```python
import itertools
from itertools import product

def CQ(Model,Query):
    '''Returns a generator of dicts, where each dict is a bijective mapping from  nodes in Model
    to all nodes in Query.
    The keys are the model nodes.
    The set of dict values equals the set of nodes of Query.
    The generator returns all answers to the Query on this Model.'''

    len_query = len(Query.nodes())
    model_nodes = list(Model.nodes)
    perms = list(itertools.product(model_nodes, repeat=len_query))
    counter = 0
    e_dict = dict()
    for n in perms:
        subgraph = Model.subgraph(n)
        if nx.is_isomorphic(subgraph, Query) is True:
            counter += 1
            query_nodes = list(Query.nodes())
            keys = list(n)
            for i in range(len(keys)):
                x = query_nodes[i]
                y = keys[i]
                e_dict[y] = x
            yield e_dict
        e_dict = dict()

# test
print(len(list(CQ(G,A))))
for n in CQ(G,A):
    print(n)
```
