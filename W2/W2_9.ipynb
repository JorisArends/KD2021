## 1 De gebruikelijke familierelaties

* We werken met predicaatlogische modellen, en met een basistaal met 2 eigenschappen: $M$ en $V$, die staan voor de 2 traditionele geslachten, en 1 twee plaatsige relatie $R$, die staat voor *is ouder van*. Hieronder staat dus dat *piet een man is, en joke een vrouw en dat piet de ouder is van joke*:

$$M(piet), V(joke), R(piet,joke).$$
* Wij gaan nu 2 dingen doen:
    1. beperkingen opleggen aan de relaties
    2. nieuwe relaties definieren in termen van de 3 primtieve relaties $M,V,R$
        * in die nieuwe relaties kan je natuurlijk andere eerder gedefinieerde relaties weer gebruiken.
* **Voorbeelden**
* *Beperking:* 
    * Iedereen is man of vrouw (en nooit allebei tegelijk).
    * Iedereen heeft exact 2 ouders, een man en een vrouw.
* *Nieuwe relatie*
    * $Vader(x,y)$ precies dan en slechts dan alleen als x een ouder is van y en x een man is.
        * $\forall xy( Vader(x,y) \leftrightarrow R(x,y) \wedge M(x))$
    * $Opa(x,y)$ hiervoor gebruik je natuurlijk het $\exists z$    connectief.
        * Je kunt dit netjes definieren in termen van ouder en vader.

Bij het lezen van het artikel van Linda kwamen we tot de conclusie dat LHBTAQI+ bestaat uit genders en sexuele orientaties. Om deze allemaal correct te kunnen definieren hebben we een aantal keuzes en aannames moeten maken om dit mogelijk te maken:
- Naast een man en een vrouw werken we met een 3e gender: transgender. Voor dit model zien we transgender als een 3e geslacht, we maken dus geen onderscheidt tussen mannen die zich een vrouw voelen, mannen die zich hebben omgebouwd tot een vrouw, vrouwen die zich een man voelen, vrouwen die zich hebben laten ombouwen tot een man en eventueel andere genders die tussen man en vrouw zitten. Ook hebben we de aanname gedaan dat een transgender persoon alleen bisexueel, pansexueel en asexueel kan zijn. Een transgender kan dus niet homo, lesbisch of hetero zijn, daarvoor moet je namelijk in ons model man of vrouw zijn.
- Een intersexueel persoon valt onder man of vrouw, afhankelijk van bij welk van de 2 dit persoon het meest 'dichtbij' zit. Als iemand bijvoorbeeld 90% man is en 10% vrouw wordt die persoon in ons model als volledig man gezien, wat ook overeenkomt de werkelijkheid aangezien een relatie van zo'n persoon met een vrouw nog als heterosexueel gezien zal worden. Als een persoon echter van mening is dat die te veel in het midden zit om als volledig man of vrouw gezien te worden, kan die beschouwd worden als transgender, waardoor die persoon dus de beperkingen zal hebben zoals beschreven in het punt hierboven.
- Een pansexueel persoon valt volgens ons model op iedereen.
- Een queer of questioning persoon voelt zich man noch vrouw noch transgender en heeft geen specifieke sexuele voorkeur. In ons model wordt deze groep gezien als mensen die geen gender hebben, dus niet man, niet vrouw, en niet transgender zijn. Ook hebben ze geen sexuele voorkeur, dus ze vallen niet op mannen, vrouwen en transgenders. Een queer of questioning is volgens ons model dus altijd asexueel, en kan dus onmogelijk aan een van de andere sexuele orientaties voldoen. Deze keuze hebben we gemaakt om de volgende reden: omdat het gender wordt geinterpreteerd als geen van drie genders, willen we de sexuele orientatie op dezelfde manier interpreteren voor uniformiteit, namelijk geen van de sexuele voorkeuren, dus asexueel. Met een kleine aanpassen zou een queer of questioning persoon wel in ons model een sexuele voorkeur kunnen hebben, als we namelijk bij bisexueel en pansexueel niet zouden 'checken' dat x een gender heeft zou die aan deze sexuele orientaties kunnen voldoen.
- Volgens ons model is de enige sexuele orientatie die op een queer of questioning valt een pansexueel persoon
- Volgens ons model valt alles wat onder de + valt in LHBTAQI+ onder een queer of questioning persoon, dus een persoon zonder een specifieke gender en sexuele voorkeur.
- Volgens ons model kan een man hetero, homo, bisexueel, pansexueel en asexueel zijn.
- Volgens ons model kan een vrouw hetero, lesbisch, bisexueel, pansexueel en asexueel zijn.
- In ons model kan een persoon niet op zichzelf vallen.
- In ons model kan een persoon niet meerdere genders tegelijk zijn. Dus een persoon kan dus niet een man en een transgender tegelijkertijd zijn. Wel kan een persoon dus helemaal geen gender hebben, namelijk queer of questioning personen en personen die vallen onder de + van LHBTAQI+


Primitieve Relaties:
* M(x): x is een man
* V(x): x is een vrouw
* T(x): x is transgender.
* R(x,y): x valt op y (x zou in potentie verliefd kunnen worden op y)
* ¬M(X)∧¬V(X)∧¬T(X): x is queer of questioning



Hieronder staat onze eerste versie van een formule die LHBTAQI+ beschrijven. Later kwamen we er echter achter dat **er tegenvoorbeelden te verzinnen waren waarbij het niet klopt**, voornamelijk in het geval van een persoon met een sexuele orientatie waar onder andere de gender die we proberen te bewijzen in zit:
* $\forall xy (Hetero(x) \leftrightarrow (R(x,y) \wedge ((M(x) \wedge V(y)) \vee (V(x) \wedge M(y)))))$  (klopt niet als x bi is)
* $\forall xy (Homo(x) \leftrightarrow (R(x,y) \wedge M(x) \wedge M(y)))$  (klopt niet als x bi is)
* $\forall xy (Lesbisch(x) \leftrightarrow (R(x,y) \wedge V(x) \wedge V(y)))$  (klopt niet als x bi is)
* $\forall xyz (Bisexueel(x) \leftrightarrow (R(x,y) \wedge M(y)) \wedge (R(x,z) \wedge V(z))$  (klopt niet als x bi is en y en z een man zijn) 


Uiteindelijk hebben we wel een manier gevonden die werkt, namelijk door voor iedere x, iedere y af te gaan en te kijken of de 'valt op' tussen x en het geslacht van y overeenkomst met wat verwacht wordt voor de sexuele orientatie van x. Ook moet eventueel het gender van x gecheckt worden, een man kan bijvoorbeeld nooit lesbisch zijn. Tot slot moet soms rekening gehouden worden met het feit dat theoretisch x en y dezelfde persoon zouden kunnen zijn.
* ∀𝑥(Bisexueel(𝑥)↔(∀y((M(y)→R(x,y))∧(V(y)→R(x,y))∧(T(y)→¬R(x,y))∨(x=y))∧(V(x)∨M(x)∨T(x))))
* ∀𝑥(Homo(𝑥)↔(∀y((M(y)→R(x,y))∧(V(y)→¬R(x,y))∧(T(y)→¬R(x,y))∨(x=y))∧M(x)))
* ∀𝑥(Lesbisch(𝑥)↔(∀y((V(y)→R(x,y))∧(M(y)→¬R(x,y))∧(T(y)→¬R(x,y))∨(x=y))∧V(x)))
* ∀𝑥(Pansexueel(𝑥)↔(∀y(R(x,y)∨(x=y))∧(V(x)∨M(x)∨T(x))))
* ∀𝑥(Asexueel(𝑥)↔¬∃y(R(x,y)))
* ∀𝑥(Hetero(𝑥)↔((∀y((M(y)→R(x,y))∧(V(y)→¬R(x,y))∧(T(y)→¬R(x,y)))∧V(x))∨(∀y((V(y)→R(x,y))∧(M(y)→¬R(x,y))∧(T(y)→¬R(x,y)))∧M(x))))

```python
# gender en sexuele orientatie netwerk
gens= nx.DiGraph()

# alle mensen met hun gender en sexuele orientatie in het netwerk
# m=man, v=vrouw, t=transgender
# Julia is een intersexueel persoon in dit netwerk, ze is 90% vrouw en 10% man,
# zoals hierboven beschreven wordt ze dus gezien als een vrouw
# Sara is queer of questioning, dus in ons model genderloos en asexueel
# Stefanie is valt onder de + van LHBTQI+, dus in ons model genderloos en asexueel
gens.add_node('Piet', M=True, V=False, T=False, S='onbekend') #bisexueel
gens.add_node('Klaas', M=True, V=False, T=False, S='onbekend') #hetero
gens.add_node('Timo', M=True, V=False, T=False, S='onbekend') #homo
gens.add_node('Liesje', M=False, V=True, T=False, S='onbekend') #lesbisch
gens.add_node('Anne', M=False, V=True, T=False, S='onbekend') #pansexueel
gens.add_node('Julia', M=False, V=True, T=False, S='onbekend') #hetero
gens.add_node('Nienke', M=False, V=True, T=False, S='onbekend') #asexueel
gens.add_node('John', M=False, V=False, T=True, S='onbekend') #pansexueel
gens.add_node('Peter', M=False, V=False, T=True, S='onbekend') #bisexueel
gens.add_node('Duncan', M=False,V=False, T=True, S='onbekend') #asexueel
gens.add_node('Sarah', M=False, V=False, T=False, S='onbekend') #asexueel
gens.add_node('Stefanie', M=False, V=False, T=False, S='onbekend') #asexueel


# nu voegen we de relatie R(x, y) toe, dus of persoon x valt op persoon y
# Piet is bisexueel
gens.add_edge('Piet','Klaas')
gens.add_edge('Piet','Timo')
gens.add_edge('Piet','Liesje')
gens.add_edge('Piet','Anne')
gens.add_edge('Piet','Julia')
gens.add_edge('Piet','Nienke')

# Klaas is hetero
gens.add_edge('Klaas','Liesje')
gens.add_edge('Klaas','Anne')
gens.add_edge('Klaas','Julia')
gens.add_edge('Klaas','Nienke')

# Timo is homo
gens.add_edge('Timo','Piet')
gens.add_edge('Timo','Klaas')

# Liesje is lesbisch
gens.add_edge('Liesje','Anne')
gens.add_edge('Liesje','Julia')
gens.add_edge('Liesje','Nienke')

# Anne is pansexueel
gens.add_edge('Anne','Piet')
gens.add_edge('Anne','Klaas')
gens.add_edge('Anne','Timo')
gens.add_edge('Anne','Liesje')
gens.add_edge('Anne','Julia')
gens.add_edge('Anne','Nienke')
gens.add_edge('Anne','John')
gens.add_edge('Anne','Peter')
gens.add_edge('Anne','Duncan')
gens.add_edge('Anne','Sarah')
gens.add_edge('Anne','Stefanie')

# Julia is hetero
gens.add_edge('Julia','Piet')
gens.add_edge('Julia','Klaas')
gens.add_edge('Julia','Timo')

# Nienke is asexueel, dus valt op niks

# John is pansexueel
gens.add_edge('John','Piet')
gens.add_edge('John','Klaas')
gens.add_edge('John','Timo')
gens.add_edge('John','Liesje')
gens.add_edge('John','Anne')
gens.add_edge('John','Julia')
gens.add_edge('John','Nienke')
gens.add_edge('John','Peter')
gens.add_edge('John','Duncan')
gens.add_edge('John','Sarah')
gens.add_edge('John','Stefanie')

# Peter is bisexueel
gens.add_edge('Peter','Piet')
gens.add_edge('Peter','Klaas')
gens.add_edge('Peter','Timo')
gens.add_edge('Peter','Liesje')
gens.add_edge('Peter','Anne')
gens.add_edge('Peter','Julia')
gens.add_edge('Peter','Nienke')

# Duncan is asexueel, dus valt op niks

# Sarah is asexueel, dus valt op niks

# Stefanie +, dus valt op niks


# dit zijn de functies waarmee we makkelijk geslacht en het bestaan van 'valt op' checken
def r(x,y,network=gens):
    return network.has_edge(x,y)
def m(x, network=gens):
    return network.nodes[x]['M']
def v(x, network=gens):
    return network.nodes[x]['V']
def t(x, network=gens):
    return network.nodes[x]['T']



# vervolgens gaan we de sexuele voorkeur van ieder persoon afleiden
# gebaseerd op het gender en op wie ze wel en niet vallen.
# doordat onze logica bestaat uit veel if statements leek het ons
# onmogelijk om er een leesbare list comprehension van te maken,
# vandaar is het zoals beneden te zien is. Wel hebben we geprobeerd
# het zoveel mogelijk te laten lijken op de manier waarop de logica is opgeschreven.
for x in gens.nodes():
    
    # bisexueel
    
    # in deze lijst moet alleen maar True en dus geen False komen te staan, dan klopt
    # het 'for every y' stuk en voldoet x aan de logica van de sexuele orientatie
    true_list_bi=[]
    for y in gens.nodes(): 
        result=True
        if x != y:
            if m(x) or v(x) or t(x):
                if m(y): 
                    result = r(x,y)
                if v(y):
                    result = r(x,y)
                if t(y):
                    result = not r(x,y)
            else:
                result=False    
            true_list_bi.append(result)
    if all(true_list_bi):
        gens.nodes[x]['S']='bisexueel'
        
        
    # homo
    true_list_homo=[]
    for y in gens.nodes():
        result=True
        if x != y:
            if m(x):
                if m(y): 
                    result = r(x,y)
                if v(y):
                    result = not r(x,y)
                if t(y):
                    result = not r(x,y)
            else:
                result=False    
            true_list_homo.append(result)
    if all(true_list_homo):
        gens.nodes[x]['S']='homo'
        
        
    # lesbisch
    true_list_les=[]
    for y in gens.nodes():
        result=True
        if x != y:
            if v(x):
                if m(y): 
                    result = not r(x,y)
                if v(y):
                    result = r(x,y)
                if t(y):
                    result = not r(x,y)
            else:
                result=False    
            true_list_les.append(result)
    if all(true_list_les):
        gens.nodes[x]['S']='lesbisch'
        
    
    # pansexueel
    true_list_pan=[]
    for y in gens.nodes():
        result=True
        if x != y:
            if m(x) or v(x) or t(x):
                result = r(x,y)         
            else:
                result=False    
            true_list_pan.append(result)
    if all(true_list_pan):
        gens.nodes[x]['S']='pansexueel'
        
    # asexueel
    true_list_asex=[]
    for y in gens.nodes():
        result=True
        if x != y:
            result = not r(x,y)           
            true_list_asex.append(result)
    if all(true_list_asex):
        gens.nodes[x]['S']='asexueel'
        
    # hetero
    true_list_het=[]
    for y in gens.nodes():
        result=True
        if x != y:
            if m(x):
                if m(y): 
                    result = not r(x,y)
                if v(y):
                    result = r(x,y)
                if t(y):
                    result = not r(x,y)
            elif v(x):
                if m(y): 
                    result = r(x,y)
                if v(y):
                    result = not r(x,y)
                if t(y):
                    result = not r(x,y)
            else:
                result=False   
            true_list_het.append(result)
    if all(true_list_het):
        gens.nodes[x]['S']='hetero'

# in de graph maken we mannen blauw, vrouwen rood, transgenders paars
# en queer of questioning en wat onder + zit grijs
kleuren = []
for node in gens.nodes():
    if gens.nodes[node]['M']==True:
        kleuren.append('blue')
    elif gens.nodes[node]['V']==True:
        kleuren.append('red')
    elif gens.nodes[node]['T']==True:
        kleuren.append('purple')
    else:
        kleuren.append('grey')
        
# bij iedere node laten we de sexuele orientatie zien
orientatie = nx.get_node_attributes(gens, 'S') 

nx.draw_circular(gens, labels=orientatie, node_color=kleuren)
```
