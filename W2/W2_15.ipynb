# Familierelaties die niet kunnen met de huidge predicaat logica
* We hebben gekozen om aangetrouwde relalties niet op te nemen. Dit omdat het niet mogelijk is om een aangetrouwde oom of tante te kunnen defineren als deze geen kinderen hebben die dan onze neven of nichten zijn. Met de huidige predicaten is het niet mogelijk om oom of tante volledig te defineren want dit kan alleen als er ook een predicaat relatie is. Een oom kan namelijk zijn als deze de broer is van een van onze ouders of een relatie heeft met een van onze tantes. 


* stel G is een twee plaatsige relatie en staat voor _is getrouwd met_ dan:
    * $Oom(x,y) $ precies dan en slechts dan alleen als x een man is en er is een w en een z en w is ouder van y en z is ouder van w en z is ouder van x of er is een z en w en een q en w is ouder van y en z is ouder van w en z is ouder van q en q is getrouwd met x
    *$\forall xy( Oom(x,y) \leftrightarrow M(x) \wedge (\exists(w,z)(R(w,y) \wedge R(z,w) \wedge R(z,x)) \vee \exists(z,w,q)(R(w,y) \wedge R(z,w) \wedge R(z,q) \wedge G(q,x)))$
    
    Dit geldt ook voor tante als x een vrouw is in plaats van een man. Omdat we de relatie _getrouwd met_ niet mogen gebruiken in de opdracht hebben we er voor gekozen om alleen oom of tante te gebruiken wanneer deze dezelfde ouders hebben als onze vader of moeder. Dit hebben we zo eerder gedefineerd in de opdracht.
    

* Een halfbroer is in onze relaties gelijk aan een broer. Dit omdat we alleen maar checken met één ouder.


* Een geadopteerd kind kun je niet defineren in de predicaat logica omdat dit kind niet woont bij zijn of haar biologische ouders.


* Een stiefbroer of zus kun je ook niet defineren in de predicaat logica omdat de stiefbroer of zus met zijn stiefbroers of zussen  geen ouder delen.


* Twee lesbische moeders kunnen in het model geen kinderen hebben / krijgen. Dit omdat ze niet allebei de biologiosche ouders zijn van het kind. Dit geldt ook voor een homo stel.

```python
G = nx.DiGraph()
Man = {1,3,6,7,9,10,11,13,15,18,19,20,22,23,24,26,27,28,30,32,34,37,39,40,43,44,46,48,49,50,52,55,56}
Vrouw = {2,4,5,8,12,14,16,17,21,25,29,31,33,35,36,38,41,42,45,47,51,53,54,57}

G.add_nodes_from(Man)
G.add_nodes_from(Vrouw)

for n in Man:
    G.nodes[n]['M']=True
    G.nodes[n]['V']=False
for n in Vrouw:
    G.nodes[n]['V']=True
    G.nodes[n]['M']=False
    
Stijn = [[3,2],[3,1],[4,1],[4,2],[8,5],[8,4],[5,6],[5,7],[9,5],[9,4],[11,10],[11,3],[12,10],[12,3],[10,52],[10,53],[10,54]]
Rico = [[15,13],[15,14],[16,13],[16,14],[22,15],[22,23],[21,15],[21,23],[24,16],[25,16],[21,17],[22,17],[21,18],[22,18],[17,19],[17,20]]
Daan = [[30,26],[30,27],[30,28],[30,29],[31,26],[31,27],[31,28],[31,29],[32,30],[33,30],[34,31],[35,31],[35,36],[34,36],[36,37],[36,38],[36,39],[55,56],[55,57],[34,55],[35,55]]
Sam = [[44,40],[44,41],[44,42],[44,43],[45,40],[45,41],[45,42],[46,45],[46,48],[47,45],[47,48],[48,49],[48,50],[48,51]]

G.add_edges_from(Stijn + Rico + Daan + Sam)

plt.figure(figsize=(16,4)) 
pos = graphviz_layout(G, prog="dot")
nx.draw(G, pos, with_labels=True, node_size=800, font_color="white", font_size=15)
plt.show()


def M(x, network=G):
    return network.nodes[x]['M']
def V(x, network=G):
    return network.nodes[x]['V']
def R(x,y,network=G):
    return network.has_edge(x,y)
```

```python
# Alle relatie comprehensies

Zoons = {(x,y) for x in G for y in G if R(y,x,G) and M(x,G)}

Dochters = {(x,y) for x in G for y in G if R(y,x,G) and V(x,G)}

Broers = {(x,y) for x in G for y in G if x != y and {z for z in G if (x,z) in Zoons and R(z,y,G) }}

Zussen = {(x,y) for x in G for y in G if x != y and {z for z in G if (x,z) in Dochters and R(z,y,G) }}

Vaders = {(x,y) for x in G for y in G if R(x,y,G) and M(x,G)}

Moeders = {(x,y) for x in G for y in G if R(x,y,G) and V(x,G)}

Opas = {(x,y) for x in G for y in G for z in G if (x,z) in Vaders and R(z,y,G)}

Omas = {(x,y) for x in G for y in G for z in G if (x,z) in Moeders and R(z,y,G)}

Tantes = {(x,y) for x in G for y in G for z in G 
         if (x,z) in Zussen and R(z,y,G)}
 
Ooms = {(x,y) for x in G for y in G for z in G
        if (x,z) in Broers and R(z,y,G)}

Neven = {(x,y) for x in G for y in G for z in G
           if (x,z) in Zoons and ((z,y) in Ooms or (z,y) in Tantes)}

Nichten = {(x,y) for x in G for y in G for z in G
            if (x,z) in Dochters and ((z,y) in Ooms or (z,y) in Tantes)}
```

```python
# Hulpfunctie voor maken van grafieken om relaties te checken
def check_graph_straight(x):
    
    T = nx.DiGraph()
    
    T.add_edges_from(x)
        
    plt.figure(figsize=(16,4)) 
    pos = graphviz_layout(G, prog="dot", root=list(x)[1])
    nx.draw(T, pos, with_labels=True, node_size=800, font_color="white", font_size=15)
    
def check_graph_curved(x):
    
    T = nx.DiGraph()
    
    T.add_edges_from(x)
        
    plt.figure(figsize=(16,8)) 
    pos = graphviz_layout(G, prog="dot")
    nx.draw(T, pos, with_labels=True, node_size=800, font_color="white", font_size=15, connectionstyle="arc3,rad=1.2")
```

```python
# Alle relaties in een grafiek om te checken of het klopt

Alle_relaties_multi = [Zoons,Dochters,Vaders,Moeders,Opas,Omas,Ooms,Tantes]
Alle_relaties_multi_naam = ["Zoons","Dochters","Vaders","Moeders","Opas","Omas","Ooms","Tantes","Neven","Nichten"]
Alle_relaties = [Broers,Zussen,Neven,Nichten]
Alle_relaties_naam = ["Broers","Zussen","Neven", "Nichten"]

for i, relatie in enumerate(Alle_relaties_multi):
    print(Alle_relaties_multi_naam[i])
    check_graph_straight(relatie)
    plt.show()

for i, relatie in enumerate(Alle_relaties):
    print(Alle_relaties_naam[i])
    check_graph_curved(relatie)
    plt.show()
```

# Uitleg Definities Letters

We hebben gekozen om de relatie R te verstaan als _aangetrokken tot_. Dit betekent niet dat iemand in R dus een relatie heeft maar dat diegene aangetrokken is tot de andere persoon. R(x,y) zegt dus dat x aangetrokken is tot y. _aangetrokken tot_ gaat niet om de persoon zelf maar meer het geslacht dat deze persoon heeft. Stel x is een man en y is een vrouw en R(x,y) dan betekent dit dat een man zich aangetrokken voelt tot het geslacht vrouw. 
Daarnaast hebben we een extra primitief _non binair_ toegevoegd. Dit betekent dat deze persoon zich niet identificeert met het geslacht man of een vrouw maar iets anders.
