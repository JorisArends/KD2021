3. Voor de duidelijkheid is bij de voorbeelden in logica steeds $\forall xy$ weggelaten.

* Vaders en Moeders
    * Niemand is zijn eigen ouder (bijv. $\neg Vader(x,x)$)
    * Ouder-relaties zijn assymetrisch (bijv. $\neg (Vader(x,y) \wedge Vader(y,x))$)
* Zoons en Dochters
    * Niemand is zijn eigen kind (bijv. $\neg Zoon(x,x)$)
    * Kind-relaties zijn assymetrisch (bijv. $\neg (Zoon(x,y) \wedge Zoon(y,x))$)
* Opa's en Oma's
    * Niemand is zijn eigen grootouder (bijv. $\neg Opa(x,x)$)
    * Groutouder-relaties zijn assymetrisch (bijv. $\neg (Opa(x,y) \wedge Opa(y,x))$)
* Kleinzoons en Kleindochters
    * Niemand is zijn eigen kleinkind (bijv. $\neg Kleinzoon(x,x)$)
    * Kleinkind-relaties zijn assymetrisch (bijv. $\neg (Kleinzoon(x,y) \wedge Kleinzoon(y,x))$)
* Broers en Zussen (en Siblings)
     * Niemand is zijn eigen Sibling (bijv. $\neg Sibling(x,x)$)
     * Als je Broer/Zus bent van iemand ben je ook gelijk Sibling (van elkaar) (dus $(Sibling(x,y) \leftrightarrow (Broer(x,y) \vee Zus(x,y)))$)
     * Sibling-relaties zijn symmetrisch (bijv. $(Sibling(x,y) \leftrightarrow Sibling(y,x) \wedge x \neq y)$)
     * Broer/zus-relaties zijn slechts symmetrisch bij hetzelfde geslacht (dus: $(ElkaarsBroer(x,y) \leftrightarrow (Broer(x,y) \wedge x \neq y \wedge M(x) \wedge M(y))$)
* Ooms en Tantes
    * Niemand is zijn eigen oom/tante (bijv. $\neg Oom(x,x)$)
    * Oom/tante-relaties zijn assymetrisch (bijv. $\neg (Oom(x,y) \wedge Oom(y,x))$)
* Nephews en Nieces
    * Niemand is zijn eigen nephew/niece (bijv. $\neg Nephew(x,x)$)
    * Nephew/niece-relaties zijn assymetrisch (bijv. $\neg (Nephew(x,y) \wedge Nephew(y,x))$)
* Cousins
    * Niemand is zijn eigen cousin (bijv. $\neg Cousin(x,x)$)
    * Cousin-relaties zijn symmetrisch (bijv. $(Cousin(x,y) \leftrightarrow Cousin(y,x) \wedge x \neq y)$)
* Neven en Nichten
    * Niemand is zijn eigen neef/nicht (bijv. $\neg Neef(x,x)$)
    * Als je Nephew/Niece bent is de relatie assymetrisch, als je Cousin bent symmetrisch.
    * Behalve als er sprake is van incest ben je slechts elkaars Cousin óf Nephew/Niece (bijv. $Neef(x,y) \leftrightarrow ((Cousin(x,y) \wedge M(x)) \oplus Nephew(x,y))$)





```python
import matplotlib.pyplot as plt

# Graaf
H = nx.DiGraph()

# Edges per groepslid
robert_edges = [(1,6),(1,8),(1,9),(2,6),(2,8),(2,9),(3,10),(4,10),(5,11),(6,11),(7,12),(7,13),(7,14),(8,12),(8,13),(8,14),(9,15),(9,16),(10,15),(10,16)]
suus_edges = [(20,22),(20,23),(19,22),(19,23),(19,17),(19,21),(18,17),(18,21),(35,18),(35,24),(35,29),(35,32), (35,33),(24,26),(24,27),(24,28),(25,26),(25,27),(25,28),(29,31),(30,31)]
nick_edges = [(36,41),(36,43),(36,44),(37,41),(37,43),(37,44),(38,45),(39,45),(40,46),(41,46),(42,47),(42,48),(42,49),(43,47),(43,48),(43,49),(44,50),(44,51),(45,50),(45,51)]
all_edges = robert_edges + suus_edges + nick_edges
H.add_edges_from(all_edges)

# Geslachten per groepslid
robert_m = {1,3,6,7,9,11,12,13,15,16}
robert_v = {2,4,5,8,10,14} 
suus_m = {19,23,24,26,27,30,31,33} 
suus_v = {17,18,20,21,22,25,28,29,32,35}
nick_m = {36,38,41,42,44,46,47,48,50,51}
nick_v = {37,39,40,43,45,49}

all_M = robert_m | suus_m | nick_m
all_V = robert_v | suus_v | nick_v

# Toewijzen
for n in all_M:
    H.nodes[n]['M']=True
    H.nodes[n]['V']=False
for n in all_V:
    H.nodes[n]['V']=True
    H.nodes[n]['M']=False
```

```python
# Primitieven
def M(x, network):
    return network.nodes[x]['M']
def V(x, network):
    return network.nodes[x]['V']
def R(x,y,network):
    return network.has_edge(x,y)

# Predicaten
Vader = {(x,y) for x in H for y in H 
    if R(x,y,H) and M(x,H)}
Moeder = {(x,y) for x in H for y in H 
    if R(x,y,H) and V(x,H)}
Zoon = {(x,y) for x in H for y in H 
    if R(y,x,H) and M(x,H)}
Dochter = {(x,y) for x in H for y in H 
    if R(y,x,H) and V(x,H)}
Opa = {(x,y) for x in H for y in H 
    if {z for z in H if R(z,y,H) and R(x,z,H)} and M(x,H)}
Oma = {(x,y) for x in H for y in H 
    if {z for z in H if R(z,y,H) and R(x,z,H)} and V(x,H)}
Kleinzoon = {(x,y) for x in H for y in H 
    if {z for z in H if R(z,x,H) and R(y,z,H)} and M(x,H)}
Kleindochter = {(x,y) for x in H for y in H 
    if {z for z in H if R(z,x,H) and R(y,z,H)} and V(x,H)}
Sibling = {(x,y) for x in H for y in H 
    if {z for z in H if R(z,x,H) and R(z,y,H)} and x!=y}
Broer = {(x,y) for x in H for y in H 
    if (x,y) in Sibling and M(x,H)}
Zus = {(x,y) for x in H for y in H 
    if (x,y) in Sibling and V(x,H)}
Oom = {(x,y) for x in H for y in H 
    if {z for z in H if R(z,y,H) and (x,z) in Broer}}
Tante = {(x,y) for x in H for y in H 
    if {z for z in H if R(z,y,H) and (x,z) in Zus}}
Nephew = {(x,y) for x in H for y in H 
    if {z for z in H if R(z,x,H) and (z,y) in Sibling} and M(x, H)}
Niece = {(x,y) for x in H for y in H 
    if {z for z in H if R(z,x,H) and (z,y) in Sibling} and V(x, H)}
Cousin = {(x,y) for x in H for y in H 
    if {z for z in H if R(z,y,H) and ((z,x) in Tante or (z,x) in Oom)}}
Neef = {(x,y) for x in H for y in H 
    if ((x,y) in Nephew) or ((x,y) in Cousin and M(x, H))}
Nicht = {(x,y) for x in H for y in H 
    if ((x,y) in Niece) or ((x,y) in Cousin and V(x, H))}

def draw_from_edges(edges, M, V, name):
    """Functie om uit een set edges een netwerk en een plotje te maken
       Vereist de pygraphviz module/package ("dot") voor het hiërarchisch plotten"""
    G = nx.DiGraph()
    G.add_edges_from(edges)
    
    for n in G:
        if n in M:
            G.nodes[n]['M']=True
            G.nodes[n]['V']=False
        else:
            G.nodes[n]['V']=True
            G.nodes[n]['M']=False
            
    mapping = {False: "#FC0FC0", True: "#0B54FE"}
    colors = [mapping[G.nodes[n]['M']] for n in G]
    pos = nx.nx_agraph.graphviz_layout(G, prog="dot")
    plt.figure(3,figsize=(10,5)) 
    nx.draw(G, pos=pos, with_labels=True, node_color=colors)
    
    ax = plt.gca()
    ax.set_title(name)
    plt.show()

# En voor elk predicaat een plaatje
predicates = {"De families": all_edges,
              "Vaders": Vader, "Moeders": Moeder,
              "Zoons": Zoon, "Dochters": Dochter, 
              "Opa's": Opa, "Oma's": Oma, 
              "Kleinzoons": Kleinzoon, "Kleindochters": Kleindochter,
              "Sibling": Sibling, "Broers": Broer, "Zussen": Zus, 
              "Ooms": Oom, "Tantes": Tante,
              "Nephews": Nephew, "Nieces": Niece,
              "Cousins": Cousin,
              "Neven": Neef, "Nichten": Nicht}

for name, predicate in predicates.items():
    draw_from_edges(predicate, all_M, all_V, name)
```

En dan de predicaten:

* $\forall x(Hetero(x) \leftrightarrow (\exists y( R(x,y) \wedge (M(x) \wedge V(y)) \vee (V(x) \wedge M(y))) \wedge \neg \exists z(R(x,z)  \wedge (M(x) \wedge M(z)) \vee (V(x) \wedge V(z)))))$

    * Oftewel: Lesbisch als een vrouw op een vrouw valt en niet op mannen.
* $\forall x(Biseksueel(x) \leftrightarrow (\exists y(R(x,y)) \wedge M(y)) \wedge (\exists z(R(x,z)) \wedge V(z)) \vee (\exists a(R(x,a)) \wedge Q(a)))$
    * Oftewel: Biseksueel als je valt op mannen, vrouwen of ongeacht hun genderidentiteit.
* $\forall x(Transgender(x) \leftrightarrow (((GM(x) \wedge V(x)) \vee (GV(x) \wedge M(x) )) \wedge \neg (GM(x) \wedge GV(x) ) ))$ 
    * Oftewel: Geboren met mannelijke geslachtskenmerken, identificeert als vrouw of geboren met vrouwelijke geslachtskenmerken, identificeert als man. Intersekse personen die een binaire genderidentieit kiezen tellen we niet mee onder de transgender definitie.
* $\forall x(Queer(x) \leftrightarrow (Q(x) \vee \neg Hetero(x) ))$
    * Oftewel: Queer is een **parapluterm** voor o.a. niet-hetero seksualiteiten, en mensen die zich identificeren als queer (bijv. veel gender non-binaire mensen).
* $\forall x(Intersekse(x) \leftrightarrow (GM(x) \wedge GV(x)) ) $
    * Oftewel: Iemand met zowel mannelijke als vrouwelijke lichaams/geslachtskenmerken.    
* +: Het plusje is vooralsnog voor al het onbenoemde oftewel: none of the above. Dit kan nog iets zijn als aseksualiteit, panseksualiteit of volgens problematische mensen zelfs pedofillie. Ook definities die nog moeten ontstaan kunnen volgen uit de plus. Daarom valt de plus in termen van primitieven niet goed te definiëren.    

---
* Mensen kunnen één van de volgende categorie letters exclusief zijn:
    * (Heteroseksueel,) Homoseksueel, Lesbisch of Biseksueel
* Een transpersoon kan dat natuurlijk ook zijn.
* Queer is een parapluterm, dus dat kan ook tegelijk.
* Intersekse en trans kan volgens onze definities niet tegelijk. Stel dat een persoon intersekse geboren wordt, daarna kiest (of wordt geforceerd) één identiteit aan te nemen, en daarna transisioneert kan het eigenlijk wel allebei, maar we houden alleen de huidige identiteit en geboortegeslacht bij.

```python
# Graaf
H = nx.DiGraph()

# 1 is een homoman
# 2 is een biseksuele man
# 3 is een lesbische vrouw
# 4 is een biseksuele vrouw
# 5 is een heteroseksuele transman
# 6 is een biseksuele queerpersoon
# 7 is een intersekse persoon die zich als vrouw identificeert en heteroseksueel is
H.add_edges_from([(1,2),(2,1),(2,4),(3,4),(4,2),(4,3),(5,4),(6,4),(6,2),(7,5)])

# Primitievensets
Mset = {1,2,5}
Vset = {3,4,7}
Qset = {6}
GVset = {3,4,5,7}
GMset = {1,2,6,7}

# Eigenschappen
for n in Mset:
    H.nodes[n]['M']=True
    H.nodes[n]['V']=False
    H.nodes[n]['Q']=False  
for n in Vset:
    H.nodes[n]['V']=True
    H.nodes[n]['M']=False
    H.nodes[n]['Q']=False
for n in Qset:
    H.nodes[n]['V']=False
    H.nodes[n]['M']=False
    H.nodes[n]['Q']=True 
for n in GMset:
    H.nodes[n]['GM']=True
    H.nodes[n]['GV']=False
for n in GVset:
    H.nodes[n]['GV']=True
    H.nodes[n]['GM']=False
for n in GMset&GVset:
    H.nodes[n]['GV']=True
    H.nodes[n]['GM']=True

# Wat primitieven als functie
def M(x, network):
    return network.nodes[x]['M']
def V(x, network):
    return network.nodes[x]['V']
def Q(x, network):
    return network.nodes[x]['Q']
def R(x,y,network):
    return network.has_edge(x,y)

# Vier predicaten als comprehensies, de rest is op basis van reeds gemaakte sets
Hetero = {x for x in H
    if ({y for y in H if R(x,y,H) and M(x,H) and V(y,H)} or {y for y in H if R(x,y,H) and V(x,H) and M(y,H)}) 
    and not ({y for y in H if R(x,y,H) and M(x,H) and M(y,H)} or {y for y in H if R(x,y,H) and V(x,H) and V(y,H)})}
Homo = {x for x in H
    if M(x,H) and {y for y in H if R(x,y,H) and M(y,H)} and not {y for y in H if R(x,y,H) and V(y,H)}}
Lesbisch = {x for x in H
    if V(x,H) and {y for y in H if R(x,y,H) and V(y,H)} and not {y for y in H if R(x,y,H) and M(y,H)}}
Biseksueel = {x for x in H
    if {y for y in H if R(x,y,H) and M(y,H)} and {y for y in H if R(x,y,H) and V(y,H)}}

# Printen wat wie is, op basis van code
print(f"""
Heteroseksuele nodes: {Hetero}
Homoseksuele nodes: {Homo}
Lesbische nodes: {Lesbisch}
Biseksuele nodes: {Biseksueel}
Trans nodes: {((GMset&Vset)|(GVset&Mset))-(GMset&GVset)} 
Queer nodes: {Qset|set(list(H.nodes))-Hetero}
Intersekse nodes: {GMset&GVset}
""")

def colornode(node):
    """Returnt kleur op basis van genderidentiteit (M/V/Q)"""
    if H.nodes[node]['M']:
        return "#0B54FE" # blauw
    elif H.nodes[node]['V']:
        return "#FC0FC0" # roze
    return "#FFFF00" # geel

def labelnode(node):
    """Labelt elke node met geboortekenmerken"""
    if H.nodes[node]['GM'] and H.nodes[node]['GV']:
        return "I"
    elif H.nodes[node]['GM']:
        return "GM"
    else:
        return "GV"

# De graaf tekenen
colors = [colornode(n) for n in H.nodes]  
plt.scatter([],[], c=["#0B54FE"], label="Man")
plt.scatter([],[], c=["#FC0FC0"], label="Vrouw")
plt.scatter([],[], c=["#FFFF00"], label="Genderqueer")
plt.legend(title="Genderidentiteit")
mapping = {node:str(node)+":"+labelnode(node) for node in H}
H = nx.relabel_nodes(H, mapping)
nx.draw(H, with_labels=True, node_color=colors, node_size=1000)
plt.show()
```
