## Primitieven:
1. **Man**

    * Iemand die zich voelt als geslacht vrouw
    * $V(x)$

    * $M_{geboren}(x)$
* **Geboren als vrouw**

    * $V_{geboren}(x)$
* **Voelt zich aangetrokken tot**
    * x voelt zich aangetrokken tot y
    * $R(x,y)$


## LHBTAQI+:
1. **Hetero**
    * Elke vrouw x waarvoor alle mensen y, waar x zich tot aangetrokken voelt, man zijn. En: elke man x waarvoor alle mensen y, waar x zich tot aangetrokken voelt, vrouw zijn.
    * $\forall xy( Hetero(x) \leftrightarrow (V(x) \wedge R(x,y) \wedge M(y)) \vee (M(x) \wedge R(x,y) \wedge V(y))$
    
* **Lesbienne**
    * Elke vrouw x waarvoor alle mensen waar x zich tot aangetrokken voelt vrouw zijn
    * $\forall xy( Lesbienne(x) \leftrightarrow (V(x) \wedge R(x,y) \wedge V(y))$
    
* **Homo**
    * Elke man x waarvoor alle mensen waar x zich tot aangetrokken voelt man zijn
    * $\forall xy( Homo(x) \leftrightarrow (M(x) \wedge R(x,y) \wedge M(y))$
    
* **Bisexueel**
    * Elk mens x dat zich zowel aangetrokken voelt tot een man y als een vrouw z
    * $\forall x( Bisexueel(x) \leftrightarrow \exists yz( R(x,y) \wedge M(y) \wedge R(x,z) \wedge V(z))$
    
* **Transgender**
    * Elke vrouw x die geboren is als man en man x die geboren is als vrouw
    * $\forall x( Transgender(x) \leftrightarrow (V(x) \wedge M_{geboren}(x)) \vee (M(x) \wedge V_{geboren}(x))$
    
* **A-sexueel**
    * Elk mens x dat zich tot niemand aangetrokken voelt
    * $\forall x( Asexueel(x) \leftrightarrow \neg\exists y(R(x,y)))$
    
* **Questioning**
    * Elk mens x dat zich niet specifiek man of vrouw voelt, noch voorkeur voelt voor een specifiek geslacht (bisexueel)
    * $\forall x( Questioning(x) \leftrightarrow (\neg V(x) \wedge \neg M(x) \wedge Bisexueel(x))$
    
* **Intersexueel**
    * Elk mens x dat niet als vrouw, noch als man is geboren
    * $\forall x( Intersexueel(x) \leftrightarrow (\neg V_{geboren}(x) \wedge \neg M_{geboren}(x))$
    
* **+**
    * Elk mens x dat niet onder de bovengenoemde categorieën valt
    * $\forall x( Plus(x) \leftrightarrow \neg (Hetero(x) \vee Lesbienne(x) \vee Homo(x) \vee Bisexueel(x) \vee Transgender(x) \vee Asexueel(x) \vee Questioning(x) \vee Intersexueel(x)))$
