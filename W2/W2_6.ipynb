###Logica

VADER: $\forall xy( Vader(x,y) \leftrightarrow Ouder(x,y) \wedge M(x))$

MOEDER: $\forall xy( Moeder(x,y) \leftrightarrow Ouder(x,y) \wedge V(x))$

BROER: $\forall xy( Broer(x,y) \leftrightarrow \exists z Vader(q,x) \wedge Vader(q,y) \wedge Moeder(z,x) \wedge Moeder(z,y) \wedge M(x))$

ZUS: $\forall xy( Zus(x,y) \leftrightarrow \exists z Vader(q,x) \wedge Vader(q,y) \wedge Moeder(z,x) \wedge Moeder(z,y) \wedge V(x))$

TANTE: $\forall xy( Tante(x,y) \leftrightarrow \exists z Zus(x,z) \wedge (Vader(z,y) \vee Moeder(z,y)) \wedge V(x))$

OOM: $\forall xy( Oom(x,y) \leftrightarrow \exists z  Broer(x,z) \wedge (Vader(z,y) \vee Moeder(z,y)) \wedge M(x))$

NEEF: $\forall xy( Neef(x,y) \leftrightarrow \exists z (Broer(q,y) \vee Zus(q,y) \wedge Zoon(x,q)) \vee \exists z (Oom(z,y) \vee Tante(z,y) \wedge Zoon(x,z))$

NICHT: $\forall xy( Nicht(x,y) \leftrightarrow \exists z (Broer(q,y) \vee Zus(q,y) \wedge Dochter(x,q)) \vee \exists z (Oom(z,y) \vee Tante(z,y) \wedge Dochter(x,z))$

OPA: $\forall xy( Opa(x,y) \leftrightarrow Ouder(x,y) \wedge Kind(y,z) \wedge M(x))$

OMA: $\forall xy( Oma(x,y) \leftrightarrow Ouder(x,z) \wedge Kind(y,z) \wedge V(x)$

AANGETROUWDEOOM: $\forall xy( Aangetrouwdeoom(x,y) \leftrightarrow \exists z Tante(z,y) \wedge Getrouwd(z,x)$

AANGETROUWDETANTE: $\forall xy( Aangetrouwdetante(x,y) \leftrightarrow \exists z Oom(z,y) \wedge Getrouwd(z,x)$

\$R(x,y)\$ betekent dat x een liefdesrelatie met y aan zou gaan.

We hebben alleen homo's, lesbiennes, bisexuelen en asexuelen gedefinieerd met liefdesrelaties omdat dit de enige sexuele voorkeuren zijn. Transgenders, queer, interseksuelen en plus zijn geen sexuele voorkeuren en zijn daarom niet te definieren met liefdesrelaties. Daarnaast is het onmogelijk om een liefdesrelatie te definieren voor bijvoorbeeld plus en queer, omdat deze mensen zelf niet weten wat hun voorkeuren of identiteit precies is. Transgenders worden beschouwd als het geslacht waar mee zij zichzelf defineren.

Mensen zijn altijd een van deze letters, of de plus. De plus is namelijk een 'catch-all' letter (symbool) en daarom valt alles wat niet in de andere letters valt daarbinnen. Iedereen valt dus binnen een van deze letters.

* $ Homo(x,y): \exists $ $R(x,M)$  $\wedge \leftrightarrow  R(x,M)$

* $ Lesbiennes(x,y):  \exists $ $R(x,V)$ $\wedge  \leftrightarrow R(x,V)$

* $ Bisexueel(x,y): \exists (R(x,V) \wedge  R(x,M)) \wedge (\leftrightarrow R(x,V) \wedge \leftrightarrow R(x,V) ))$ $ $

* $ Asexueel(x,y): \not \exists (R(x,V) \wedge  R(x,M)) \wedge (\not \leftrightarrow R(x,V) \wedge \not \leftrightarrow R(x,M) ))$ $ $
