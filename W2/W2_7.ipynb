## Primitieven

    * Transgender betekent dat iemand zich niet indentificeert met het geslacht waarmee hij of zij geboren is en aangezien M en V staan voor het traditionele geslachten die zich wel identificeren met het geslacht waarmee zij zijn geboren is deze primitief toegevoegd om hierin onderscheid te maken.

* De twee plaatsige liefdesrelatie $R(x,y)$ betekent dat $x$ houdt van of in het verleden heeft gehouden van $y$. 
 
## Logische relaties 
### Heteroseksueel

Heteroseksuelen worden gedefinieerd als personen die enkel van personen van het andere traditionele geslacht houden of hebben gehouden. Dit zijn dus mannen die enkel van vrouwen houden of hebben gehouden of vrouwen die enkel van mannen houden of hebben gehouden.

$\forall x \exists y( Heteroseksueel(x) \leftrightarrow (R(x,y) \wedge M(x) \wedge V(y) \wedge \neg \exists z(R(x,z) \wedge M(z))) \vee (R(x,y) \wedge V(x) \wedge M(y) \wedge \neg \exists z(R(x,z) \wedge V(z)))   $

### Lesbisch

Lesbiennes worden gedefinieerd als vrouwen die enkel en alleen houden of hebben gehouden van andere vrouwen.

$\forall x \exists y(Lesbisch(x) \leftrightarrow R(x,y) \wedge V(x) \wedge V(y) \wedge \neg \exists z(R(x,z) \wedge M(z))) $

### Homoseksueel

Homoseksuelen worden gedefinieerd als mannen die enkel en alleen houden of hebben gehouden van andere mannen.

$\forall x \exists y(Homoseksueel(x) \leftrightarrow R(x,y) \wedge M(x) \wedge M(y) \wedge \neg \exists z(R(x,z) \wedge V(z))) $


### Biseksueel

Biseksuelen worden gedefinieerd als personen die zowel van mannen als vrouwen houden of hebben gehouden.

$\forall x(Biseksueel(x) \leftrightarrow \exists y(R(x,y) \wedge V(y)) \wedge \exists z(R(x,z) \wedge M(z)))$

### Queer

Queer wordt beschouwd als een 'umbrella-term' waaronder alle personen vallen die zich identificeren met 1 of meerdere letters van het LHBTQI+ spectrum. 

$\forall x(Queer(x) \leftrightarrow Homoseksueel(x) \vee Lesbisch(x) \vee Biseksueel(x) \vee Aseksueel(x) \vee Transgender(x) \vee Intersex(x)$)

### Aseksueel +

Aseksuelen worden gedefinieerd als personen die van geen enkel ander persoon houden of hebben gehouden.

$\forall x(Aseksueel(x) \leftrightarrow \neg \exists y(R(x,y)))$

```python
H = nx.DiGraph() 

# de edges zijn de tweeplaatsige liefdesrelaties
H.add_edges_from([('Ron', 'Hermione'), ('Harry', 'Ginny'), ('Dumbledore', 'Grindelwald'), ('Bellatrix', 'Nymphadora'), ('Nymphadora', 'Bellatrix'), ('Nymphadora', 'Remus'), ('Narcissa', 'Lucius'), ('Lucius', 'Narcissa'), ('Cedric', 'Harry'), ('Luna', 'Hermione'), ('Minerva', 'Dumbledore'), ('Cho', 'Cedric'), ('Molly', 'Arthur'), ('Harry', 'Hermione')])

# de verschillende primitieven 
M,V,T,I = {'Ron', 'Harry', 'Dumbledore', 'Grindelwald', 'Remus', 'Lucius', 'Cedric', 'Arthur'}, {'Hermione', 'Ginny', 'Bellatrix', 'Nymphadora', 'Narcissa', 'Luna', 'Molly'}, {'Minerva'}, {'Cho'}

# de attributen M, V, T en I goed zetten
for n in M:
    H.nodes[n]['M']=True
    H.nodes[n]['V']=False
    H.nodes[n]['T']=False
    H.nodes[n]['I']=False
    
for n in V:
    H.nodes[n]['V']=True
    H.nodes[n]['M']=False 
    H.nodes[n]['T']=False
    H.nodes[n]['I']=False
    
for n in T: 
    H.nodes[n]['V']=False
    H.nodes[n]['M']=False 
    H.nodes[n]['I']=False
    H.nodes[n]['T']=True
    
for n in I: 
    H.nodes[n]['V']=False
    H.nodes[n]['M']=False 
    H.nodes[n]['T']=False
    H.nodes[n]['I']=True
    
    
# afkortingen voor de comprenhensies
G=H
def M(x, network=G):
    return network.nodes[x]['M']
def V(x, network=G):
    return network.nodes[x]['V']
def T(x, network=G):
    return network.nodes[x]['T']
def I(x, network=G):
    return network.nodes[x]['I']
def R(x,y,network=G):
    return network.has_edge(x,y)

# add colours to the nodes
node_color = []
for node in H.nodes(data=True):
    if node[1]['M']:
        node_color.append('blue')
    elif node[1]['V']:
        node_color.append('pink') 
    elif node[1]['T']:
        node_color.append('yellow')
    elif node[1]['I']:
        node_color.append('purple')

# nx.draw_networkx(H)
nx.draw(H, with_labels=True, node_size=120, node_color=node_color)

# de comprenhensies
Hetero = {x for x in H for y in H if R(x,y,H) and M(x,H) and V(y,H) and not {z for z in H if R(x,z,H) and M(z,H)} or R(x,y,H) and V(x,H) and M(y,H) and not {z for z in H if R(x,z,H) and V(z,H)}}
Homo = {x for x in H for y in H if R(x,y,H) and M(x,H) and M(y,H) and not {z for z in H if R(x,z,H) and V(z,H)}}
Lesbisch = {x for x in H for y in H if R(x,y,H) and V(x,H) and V(y,H) and not {z for z in H if R(x,z,H) and M(z,H)}}
Biseksueel = {x for x in H if {y for y in H if R(x,y,H) and V(y,H)} and {y for y in H if R(x,y,H) and M(y,H)}}
Trans = {x for x in H if T(x,H)}
Intersex = {x for x in H if I(x,H)}
Aseksueel = {x for x in H if not {z for z in H if R(x,z,H)}}
Queer = Homo | Lesbisch | Biseksueel | Aseksueel | Trans | Intersex
```
