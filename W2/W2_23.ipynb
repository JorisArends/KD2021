* We werken met predicaatlogische modellen, en met een basistaal met 2 eigenschappen: $M$ en $V$, die staan voor de 2 traditionele geslachten, en 1 twee plaatsige relatie $R$, die staat voor *is ouder van*. Hieronder staat dus dat *piet een man is, en joke een vrouw en dat piet de ouder is van joke*:

$$M(piet), V(joke), R(piet,joke).$$
* Wij gaan nu 2 dingen doen:
    1. beperkingen opleggen aan de relaties
    2. nieuwe relaties definieren in termen van de 3 primtieve relaties $M,V,R$

* *Beperking:*

    * $Vader(x,y)$ precies dan en slechts dan alleen als x een ouder is van y en x een man is.

    * $Opa(x,y)$ hiervoor gebruik je natuurlijk het $\exists z$    connectief.

## Vader en Moeder

Vader (ouderrelatie tussen twee personen en ouder is een man) :

$\forall xy( Vader(x,y) \leftrightarrow R(x,y) \wedge M(x))$

Moeder (ouderrelatie tussen twee personen en ouder is een vrouw):

$\forall xy( Moeder(x,y) \leftrightarrow R(x,y) \wedge V(x))$

```python
Vader = {(x,y) for x in G for y in G if R(x,y) and M(x)}

vader = maak_netwerk(Vader)
pos = graphviz_layout(vader, prog="dot")

plt.figure(figsize=(18,10))
nx.draw(vader, with_labels=True, pos=pos, node_size=600)
```

## (Half)Broer en (Half)Zus

Broer (twee personen met dezelfde vader en moeder en de broer is een man):

$\forall xy( Broer(x,y) \leftrightarrow M(x) \wedge \exists zw((Vader(w,x) \wedge Vader(w,y) \wedge (Moeder(z,x) \wedge Moeder(z,y)))$

Zus (twee personen met dezelfde vader en moeder en de zus is een vrouw):

$\forall xy( Zus(x,y) \leftrightarrow V(x) \wedge \exists zw((Vader(w,x) \wedge Vader(w,y) \wedge (Moeder(z,x) \wedge Moeder(z,y)))$

Halfbroer (twee personen met maximaal één zelfde ouder en de halfbroer is een man):

$\forall xy( Halfbroer(x,y) \leftrightarrow M(x) \wedge \exists zw((Vader(w,x) \wedge Vader(w,y) \oplus (Moeder(z,x) \wedge Moeder(z,y)))$

Halfzus (twee personen met maximaal één zelfde ouder en de halfzus is een vrouw):

$\forall xy( Halfzus(x,y) \leftrightarrow V(x) \wedge \exists zw((Vader(w,x) \wedge Vader(w,y) \oplus (Moeder(z,x) \wedge Moeder(z,y)))$

```python
Broer = {(x,y) for x in G for y in G 
        if {z for z in G if (R(z,x) and R(z,y)) and M(z)} 
        and {z for z in G if (R(z,x) and R(z,y)) and V(z)} 
        and M(x)
        and x != y}

broer = maak_netwerk(Broer)
pos = graphviz_layout(broer, prog="twopi")

plt.figure(figsize=(16,10))
nx.draw(broer, with_labels=True, pos=pos, node_size=600)
```

```python
Zus = {(x,y) for x in G for y in G 
        if {z for z in G if (R(z,x) and R(z,y)) and M(z)} 
        and {z for z in G if (R(z,x) and R(z,y)) and V(z)} 
        and V(x)
        and x != y}

zus = maak_netwerk(Zus)
pos = graphviz_layout(zus, prog="twopi")

plt.figure(figsize=(18,10))
nx.draw(zus, with_labels=True, pos=pos, node_size=600)
```

```python
Halfbroer = {(x,y) for x in G for y in G 
        if xor(len({z for z in G if (R(z,x) and R(z,y)) and M(z)}) > 0, len({z for z in G if (R(z,x) and R(z,y)) and V(z)}) > 0)
        and M(x)
        and x != y}

halfbroer = maak_netwerk(Halfbroer)
pos = graphviz_layout(halfbroer, prog="twopi")

plt.figure(figsize=(18,10))
nx.draw(halfbroer, with_labels=True, pos=pos, node_size=600)
```

```python
Halfzus = {(x,y) for x in G for y in G 
        if xor(len({z for z in G if (R(z,x) and R(z,y)) and M(z)}) > 0, len({z for z in G if (R(z,x) and R(z,y)) and V(z)}) > 0)
        and V(x)
        and x != y}

halfzus = maak_netwerk(Halfzus)
pos = graphviz_layout(halfzus, prog="twopi")

plt.figure(figsize=(18,10))
nx.draw(halfzus, with_labels=True, pos=pos, node_size=600)
```

## Opa en Oma

Opa (een derde persoon is het kind van de opa en ouder van het kleinkind):

$\forall xy( Opa(x,y) \leftrightarrow M(x) \wedge \exists z(R(x,z) \wedge R(z,y)))$

Oma (een derde persoon is het kind van de oma en ouder van het kleinkind):

$\forall xy( Oma(x,y) \leftrightarrow V(x) \wedge \exists z(R(x,z) \wedge R(z,y)))$

```python
Opa = {(x,y) for x in G for y in G
       if M(x)
       and {z for z in G if R(x,z)} & {z for z in G if R(z,y)}}

opa = maak_netwerk(Opa)
pos = graphviz_layout(opa, prog="dot")

plt.figure(figsize=(18,10))
nx.draw(opa, with_labels=True, pos=pos, node_size=600)
```

```python
Oma = {(x,y) for x in G for y in G
       if V(x)
       and {z for z in G if R(x,z)} & {z for z in G if R(z,y)}}

oma = maak_netwerk(Oma)
pos = graphviz_layout(oma, prog="dot")

plt.figure(figsize=(18,10))
nx.draw(oma, with_labels=True, pos=pos, node_size=600)
```

## Oom en Tante

Tante (moeders kant/vaders kant):

$\forall xy( Tante(x,y) \leftrightarrow V(x) \wedge \exists z(Zus(x,z) \wedge R(z,y))$

Oom (moeders kant/vaders kant):

$\forall xy( Oom(x,y) \leftrightarrow M(x) \wedge \exists z(Broer(x,z) \wedge R(z,y))$

Beperking: aangetrouwde oom/tante is niet te verkrijgen

```python
Oom = {(x,y) for x in G for y in G
         if M(x)
         and {z for z in G if (x,z) in Broer and R(z,y)}}

oom = maak_netwerk(Oom)
pos = graphviz_layout(oom, prog="dot")

plt.figure(figsize=(18,10))
nx.draw(oom, with_labels=True, pos=pos, node_size=600)
```

```python
Tante = {(x,y) for x in G for y in G
         if V(x)
         and {z for z in G if (x,z) in Zus and R(z,y)}}

tante = maak_netwerk(Tante)
pos = graphviz_layout(tante, prog="dot")

plt.figure(figsize=(18,10))
nx.draw(tante, with_labels=True, pos=pos, node_size=600)
```

## Neef en Nicht

Neef (zoon van de zus of broer van je moeder of vader):

$\forall xy( Neef(x,y) \leftrightarrow M(x) \wedge \exists z((Tante(z,y) \lor Oom(z,y)) \wedge R(z,x)))$

Nicht (dochter van de zus of broer van je moeder of vader):

$\forall xy( Nicht(x,y) \leftrightarrow V(x) \wedge \exists z((Tante(z,y) \lor Oom(z,y)) \wedge R(z,x)))$

```python
Neef = {(x,y) for x in G for y in G
          if M(x)
          and {z for z in G if ((((z,y) in Tante)) | ((z,y) in Oom)) & (R(z,x))}}

neef = maak_netwerk(Neef)
pos = graphviz_layout(neef, prog="circo")

plt.figure(figsize=(18,10))
nx.draw(neef, with_labels=True, pos=pos, node_size=600)
```

```python
Nicht = {(x,y) for x in G for y in G
          if V(x)
          and {z for z in G if ((((z,y) in Tante)) | ((z,y) in Oom)) & (R(z,x))}}

nicht = maak_netwerk(Nicht)
pos = graphviz_layout(nicht, prog="circo")

plt.figure(figsize=(18,10))
nx.draw(nicht, with_labels=True, pos=pos, node_size=600)
```

## Overgrootvader en Overgrootmoeder

Overgrootvader (Opa van je moeder of vader):

$\forall xy( Overgrootvader(x,y) \leftrightarrow M(x) \wedge \exists z(Opa(x,z) \wedge R(z,y)))$

Overgrootmoeder (Oma van je moeder of vader):

$\forall xy( Overgrootmoeder(x,y) \leftrightarrow V(x) \wedge \exists z(Oma(x,z) \wedge R(z,y)))$

```python
Overgrootvader = {(x,y) for x in G for y in G
                   if M(x)
                   and {(z,w) for z in G for w in G if R(x,z) and R(z,w) and R(w,y)}}

overgrootvader = maak_netwerk(Overgrootvader)
pos = graphviz_layout(overgrootvader, prog="dot")

plt.figure(figsize=(18,10))
nx.draw(overgrootvader, with_labels=True, pos=pos, node_size=600)
```

```python
Overgrootmoeder = {(x,y) for x in G for y in G
                   if V(x)
                   and {(z,w) for z in G for w in G if R(x,z) and R(z,w) and R(w,y)}}

overgrootmoeder = maak_netwerk(Overgrootmoeder)
pos = graphviz_layout(overgrootmoeder, prog="dot")

plt.figure(figsize=(18,10))
nx.draw(overgrootmoeder, with_labels=True, pos=pos, node_size=600)
```

## Opdracht 1.2

Familierelaties die niet in predicaatlogica kunnen worden gedefinieerd:
- Stiefbroer/zus/vader/moeder etc. Dit kan niet omdat er geen ouder-kind relatie bestaat met iemand uit jouw familie. Er is dus geen enkele manier om via predicaatlogica tot deze persoon te komen. 
- Geadopteerd kind. Wanneer je de ouder-kind relatie beschouwt als biologisch, heeft dit kind geen relaties met de pleegouders.
- Kinderloze aangetrouwde oom/tante. Aangezien de persoon geen kinderen heeft zijn er geen ouder-kind relaties binnen de familie. De persoon is ook aangetrouwd dus heeft niet dezelfde ouders als bijvoorbeeld je vader.

## Opdracht 1.3

Iedereen is man óf vrouw

$\forall x(M(x) \oplus V(x))$

Iedereen heeft twee ouders, een vader en een moeder

$\forall x \exists yz(x \rightarrow Vader(y,x) \wedge Moeder(z,x))$

Niemand is ouder van zichzelf

$\forall x(\neg R(x,x))$

Ouder is asymmetrisch

$\forall xy(R(x,y) \rightarrow \neg R(y,x))$

Primitieven:
- **(Biologisch) Man: M**
    - De identificatie van het geslacht 'man' door de fysieke geslachtskenmerken waarmee de persoon geboren wordt
- **(Biologisch) Vrouw: V**
    - De identificatie van het geslacht 'vrouw' door de fysieke geslachtskenmerken waarmee de persoon geboren wordt
- **(Identificeert als) Man: iM**
    - De manier waarop een persoon zich vanbinnen voelt over zijn/haar geslacht (in dit geval man) en hoe diegene geïdentificeerd of aangesproken wilt worden (als man)
- **(Identificeert als) Vrouw: iV**
    - De manier waarop een persoon zich vanbinnen voelt over zijn/haar geslacht (in dit geval vrouw) en hoe diegene geïdentificeerd of aangesproken wilt worden (als vrouw)
- **Zich aangetrokken voelen tot: R**
    - Twee personen zijn in relatie wanneer persoon x zich seksueel voelt aangetrokken tot persoon y.
    - Dit kan wederzijds zijn, maar dat hoeft niet.

```python
H = nx.DiGraph()

m = {'Homo1', 'Bi1', 'Interseks1', 'Queer1', 'Trans2'} 
v ={'Lesbi1', 'Lesbi2', 'Interseks1', 'Trans1', 'Pan1'}
im = {'Homo1', 'Bi1', 'Interseks1', 'Trans1'}
iv = {'Lesbi1', 'Lesbi2', 'Pan1', 'Trans2'}

H.add_nodes_from((m|v|im|iv))
for n in H:
    H.nodes[n]['bio_gender_man']=False
    H.nodes[n]['bio_gender_vrouw']=False
    H.nodes[n]['i_gender_man']=False
    H.nodes[n]['i_gender_vrouw']=False
    
for n in m:
    H.nodes[n]['bio_gender_man']=True

for n in v:
    H.nodes[n]['bio_gender_vrouw']=True
    
for n in im:
    H.nodes[n]['i_gender_man']=True
    
for n in iv:
    H.nodes[n]['i_gender_vrouw']=True
    

relaties = {('Homo1', 'Bi1'), ('Lesbi1', 'Lesbi2'), ('Bi1', 'Homo1'), 
            ('Lesbi2', 'Lesbi1'), ('Bi1', 'Lesbi2'), ('Interseks1', 'Homo1'), 
            ('Queer1', 'Lesbi2'), ('Queer1', 'Interseks1'), ('Bi1', 'Trans1'),
            ('Pan1', 'Queer1'), ('Trans2', 'Lesbi2')}

H.add_edges_from(relaties)

nx.draw(H, with_labels=True)

```

## Lesbisch

Je identificeert jezelf als vrouw en je voelt je alleen maar aangetrokken tot mensen die zich identificeren als vrouw

$\forall x( L(x) \leftrightarrow iV(x) \wedge \forall y(iV(y) \rightarrow R(x,y)))$

Trans2 wordt ook geïdentificeerd als Lesbisch, aangezien volgens onze definitie van Lesbisch iemand Lesbisch is als je jezelf identifceert als vrouw en je alleen maar aangetrokken voelt tot mensen die zich identificeren als vrouw. Trans2 is man, maar identificeert zich als vrouw. Trans2 voelt zich aangetrokken tot Lesbi2, en dat is de enige persoon waar Trans2 zich tot voelt aangetrokken. Volgens onze definitie is Trans2 dus ook Lesbisch.

```python
Lesbisch = {(x,y) for x in H 
            if H.nodes[x]['i_gender_vrouw'] 
            and not H.nodes[x]['i_gender_man'] 
            and not {z for z in H if R(x,z,H) and H.nodes[z]['i_gender_man']}
            for y in H 
            if H.nodes[y]['i_gender_vrouw']
            and not H.nodes[x]['i_gender_man'] 
            and R(x,y,H)}


nx.draw_kamada_kawai(maak_netwerk(Lesbisch), with_labels=True)
print(Lesbisch)
```

## Homo

Je identificeert jezelf als man en je voelt je alleen maar aangetrokken tot mensen die zich identificeren als man

$\forall x( H(x) \leftrightarrow iM(x) \wedge \forall y(iM(y) \rightarrow R(x,y)))$

Interseks1 wordt ook geïdentificeerd als Homo, aangezien volgens onze definitie van Homo iemand Homo is als je jezelf identifceert als man en je alleen maar aangetrokken voelt tot mensen die zich identificeren als man. Interseks1 is geen man of vrouw, maar identificeert zich als man. Interseks1 voelt zich aangetrokken tot Homo1, en dat is de enige persoon waar Interseks1 zich tot voelt aangetrokken. Volgens onze definitie is Interseks1 dus ook Homo.

```python
Homo = {(x,y) for x in H
        if H.nodes[x]['i_gender_man'] 
        and not H.nodes[x]['i_gender_vrouw'] 
        and not {z for z in H if R(x,z,H) and H.nodes[z]['i_gender_vrouw']}
        for y in H 
        if H.nodes[y]['i_gender_man'] 
        and not H.nodes[y]['i_gender_vrouw'] 
        and R(x,y,H)}

nx.draw_kamada_kawai(maak_netwerk(Homo), with_labels=True)
print(Homo)
```

## Bi

Je identificeert jezelf als man of vrouw en je voelt je aangetrokken tot minstens één iemand die zich identificeert als man en minstens één iemand die zich identificeert als vrouw

$\forall x( B(x) \leftrightarrow \exists yz((R(x,y) \wedge iM(y)) \wedge (R(x,z) \wedge iV(y))$

Bi1 is biseksueel, omdat hij/zij zowel valt op iemand die zich als man identificeert (Homo1 en Trans1) als iemand die zich als vrouw identficeert (Lesbi2)

```python
Bi = {(x,y) for x in H 
            if H.nodes[x]['i_gender_vrouw']
            or H.nodes[x]['i_gender_man']
            if {z for z in H if H.nodes[z]['i_gender_man'] and R(x,z,H)}
            and {z for z in H if H.nodes[z]['i_gender_vrouw'] and R(x,z,H)}
            for y in H
            if R(x,y,H)}

print(Bi)
nx.draw_kamada_kawai(maak_netwerk(Bi), with_labels=True)
```

## Transgender

Je bent biologisch man en identificeert jezelf als vrouw of je bent biologisch vrouw en identificeert jezelf als man

$\forall x( T(x) \leftrightarrow (M(x) \wedge iV(x)) \lor (V(x) \wedge iM(x)))$

```python
Trans = {(x,x) for x in H 
            if H.nodes[x]['bio_gender_vrouw']
            and not H.nodes[x]['bio_gender_man']
            and H.nodes[x]['i_gender_man']
            or H.nodes[x]['bio_gender_man']
            and not H.nodes[x]['bio_gender_vrouw']
            and H.nodes[x]['i_gender_vrouw']
            }

print(Trans)
nx.draw_kamada_kawai(maak_netwerk(Trans), with_labels=True)
```

## Queer

Je identificeert jezelf niet als man en niet als vrouw

$\forall x( Q(x) \leftrightarrow \neg iM(x) \wedge \neg iV(x))$

```python
Queer = {(x,x) for x in H 
            if not H.nodes[x]['i_gender_man']
            and not H.nodes[x]['i_gender_vrouw']
                }

print(Queer)
nx.draw_kamada_kawai(maak_netwerk(Queer), with_labels=True)
```

## Interseksueel

Je bent biologisch man en vrouw tegelijk omdat je bent geboren met mannelijke en vrouwelijke fysieke geslachtskenmerken

$\forall  x( I(x) \leftrightarrow M(x) \wedge V(x))$

```python
Inter = {(x,x) for x in H 
            if H.nodes[x]['bio_gender_man']
            and H.nodes[x]['bio_gender_vrouw']
                }

print(Inter)
nx.draw_kamada_kawai(maak_netwerk(Inter), with_labels=True)
```

## Plus

+: Je bent geen van bovenstaande

$\forall x( +(x) \leftrightarrow \neg L(x) \wedge \neg H(x) \wedge \neg B(x) \wedge \neg T(x) \wedge \neg Q(x) \wedge \neg I(x))$

Toelichting “Plus” ofwel Panseksueel: Een panseksueel is iemand die geen labels plakt aan waar hij/zij op valt. De persoon in kwestie valt op mannen, vrouwen en alles er tussenin zonder vooroordelen.


```python
#Moet nog gedaan worden

Plus = {(x,x) for x in H 
            if x not in {homo[0] for homo in Homo}
            if x not in {lesbi[0] for lesbi in Lesbisch}
            if x not in {bi[0] for bi in Bi}
            if x not in {trans[0] for trans in Trans}
            if x not in {queer[0] for queer in Queer}
            if x not in {inter[0] for inter in Inter}
            }

print(Plus)
nx.draw_kamada_kawai(maak_netwerk(Plus), with_labels=True)
```
