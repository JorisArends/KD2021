#### 2. Uitzonderingen
Er zijn een drietal uitzonderingen die wij hebben gevonden en dus niet weer te geven zijn met een logische zin. Er zijn veel meer variaties op mogelijk, maar dit zijn de generalisaties ervan. We zijn er hier vanuit gegaan dat alleen de primitieven M, V en R mogen worden gebruikt.
* Partner: zolang er geen kinderen zijn, is het niet mogelijk om weer te geven dat iemand een partner heeft. Als een tante getrouwd is met een man en zij geen kinderen hebben, is deze oom op logische wijze geen familie. Wanneer zij wel kinderen hebben kan de logische relatie via de neef/nicht worden weergegeven. 
* Schoonfamilie: dit sluit eigenlijk aan op dat partners niet kunnen weergegeven. Zolang een persoon en zijn/haar partner geen kinderen hebben kan ook de rest van de schoonfamilie van deze persoon niet worden weergegeven in een logisch verband.
* Stieffamilie: ook dit sluit weer aan op dat partners niet kunnen worden weergegeven als er geen kind is dat van een stiefvader en moeder is of andersom. Via een halfzus of halfbroer zou deze relatie naar de stieffamilie wel kunnen worden gelegd, omdat je dan bijvoorbeeld kan aangeven dat je wel dezelfde moeder hebt maar niet dezelfde vader. Op die manier kan je dan ook weer logische verbanden leggen naar de 
* Meerdere broers of zussen: wanneer iemand meer dan één broer of meer dan één zus heeft, kan dat niet worden aangegeven met deze primitieven, omdat twee broers dezelfde ouders en hetzelfde geslacht hebben, waardoor de logische zinnen er hetzelfde uitzien. 

#### 3. Beperkingen
Hieronder staan de beperkingen en de daarbij behorende logische zinnen die wij hebben gevonden.
* Iedereen is een man of een vrouw
    * $\forall x( M(x) \lor V(x))$
* Niemand is een man en een vrouw
    * $\lnot \forall x( M(x) \wedge V(x))$
* Niemand is ouder van zichzelf
    * $\lnot \forall x( R(x,x))$
* 'Ouder van' is asymmetrisch: als je de ouder van bent van iemand kan die persoon niet ook de ouder van jou zijn
    * $\forall xy( R(x,y) \leftrightarrow \lnot R(y,x))$
* 'Ouder van' is intransitief: wanneer je de ouder bent van iemand kan deze persoon niet ook de ouder zijn van iemand waarvan jij ook de ouder bent
    * $\forall xy( R(x,y) \leftrightarrow \lnot \exists ( R(z,y) \wedge R(x,z))$
    * Een uitzondering kan hierop voorkomen wanneer een vader een kind heeft met zijn dochter of een moeder een kind heeft met haar zoon.
* Iedereen heeft een man en een vrouw als ouder
    * $\forall xyz( R(x,y) \wedge R(z,y) \leftrightarrow (M(x) \wedge V(z)) \lor (M(z) \wedge V(x)))$
    * In de realiteit kan iemand ook twee vrouwen, twee mannen of meerdere ouders hebben.

**Primitieven:**
* Man $M$ - Mensen geboren als man die zich als man identificeren.
* Getransformeerde man $Mt$ - Mensen geboren als vrouw, maar zich nu als man identificeren.
* Vrouw $V$ - Mensen geboren als vrouw die zich als vrouw identificeren.
* Getransformeerde vrouw $Vt$ - Mensen geboren als man, maar zich nu als vrouw identificeren.
* Niet-gedefinieerd $N$ - Mensen geboren als man of vrouw die zich noch man noch vrouw voelen.
* Valt op $R(x,y)$ - x valt op y.

<br>**Onze definities van de letters:**
<br>$Lesbiennes ~ L$ - Dit zijn vrouwen $V$ die alleen op vrouwen $V$ vallen:
<br>$\forall x (L(x) \rightarrow V(x) \wedge \exists y(V(y) \wedge R(x,y))$
<br>
<br>$Homosexuelen ~ H$ - Dit zijn mannen $M$ die alleen op mannen $M$ vallen:
<br>$\forall x (H(x) \rightarrow M(x) \wedge \exists y(M(y) \wedge R(x,y))$
<br>
<br>$Bisexuelen ~ B$ - Dit zijn mannen $M$ of vrouwen $V$ die alleen op mannen $M$ en vrouwen $V$ vallen:
<br>$\forall x (B(x) \rightarrow (V(x) \vee M(x)) \wedge \exists y((V(y) \vee M(y)) \wedge R(x,y))$
<br>
<br>$Transsexuelen ~ T$ - Dit zijn getransformeerde mannen $M$ of vrouwen $V$ waarvan de voorkeur niet afhangt van de letter:
<br>$\forall x (T(x) \rightarrow (Mt(x) \vee Vt(x))$
<br>
<br>$Queer ~ Q$ - Dit zijn niet-gedefinieerden $N$ die op mannen $M$, vrouwen $V$, transgenders $Mt$/$Vt$ of niet-gedefinieerden $N$ vallen:
<br>$\forall x (Q(x) \rightarrow N(x) \wedge \exists y((M(y) \vee V(y) \vee Mt(y) \vee Vt(y)) \vee N(y) \wedge R(x,y))$
<br>
<br>$Intersexueel ~ I$ - Dit zijn niet-gedefinieerden $N$ waarvan de voorkeur niet afhangt van de letter:
<br>$\forall x(I(x) \rightarrow N(x))$
<br>
<br>$Plusteken ~ +$ - None of the above:
<br>Wij stellen dat je dit niet met de primitieven kan definiëren. Het plusteken is er voor mensen die vinden dat ze niet in het rijtje thuishoren. Volgens ons zijn er te veel mogelijkheden die niet in het rijtje horen om deze in de gestelde primitieven te definiëren. Het is zeker mogelijk voorbeelden van gevallen die onder $+$ vallen te definiëren, maar niet allemaal. Via het onderstaande model kwamen we er wel, maar daar worden de letters gebruikt en niet de primitieven. 
<br>$\forall x(+(x) \rightarrow \neg L(x) \wedge \neg H(x) \wedge \neg B(x) \wedge \neg T(x) \wedge \neg Q(x) \wedge \neg I(x))$
<br>
<br>Wij hebben gekozen voor 'exists' bij $L,H,B$ en $Q$ omdat wij ervan uitgaan dat bijvoorbeeld een lesbienne niet op *elke* vrouw valt. Mensen hebben immers andere voorkeuren dan alleen geslacht.
