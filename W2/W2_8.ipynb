* MV(x): x was een man en is nu een vrouw
* VM(x): x was een vrouw en is nu een man
* Q(x): x is Queer
* I(x): x is Intersexueel
* B(x): x is Bisexueel



Primitief:
$\forall x(Transgender(x) \iff MV(x) \vee VM(x))$

**Voor alle** x geldt dat x Transgender is **als** x een man was en nu vrouw is **of** als x een vrouw was en nu man is.
<br><br>

$\forall xy(Lesbie(x,y) \iff \exists R(x,y) \wedge (V(x) \vee MV(x)) \wedge (V(y) \vee MV(y))$

**Voor alle** x en y geldt dat x Lesbisch is **als** x nu een vrouw is en dan dus niet Q,I of B is en **als** x een relatie had/heeft gehad met y **en** x nu dus een vrouw is **en** y nu een vrouw is.
<br><br>

$\forall xy(Homo(x,y) \iff \exists R(x,y) \wedge (M(x) \vee VM(x)) \wedge (M(y) \vee MV(y))$

**Voor alle** x en y geldt dat x Homo is **als** x nu een man is en dan dus niet Q,I of B en **als** x een relatie had/heeft gehad met y **en** x nu dus een man is **en** y nu een man is.
<br><br>

$\forall xy(Interseksueel relatie(x,y) \iff \exists R(x,y) \wedge (I(x) \vee I(y))$

**Voor alle** x geldt dat x een Interseksuele relatie heeft **als** x en y een relatie hebben **en** x **of** y Interseksueel is.
<br><br>

$\forall xy(Queer relatie(x,y) \iff \exists R(x,y) \wedge Q(x) \vee Q(y) $

**Voor alle** x geldt dat x een Queer relatie heeft **als** x en y een relatie hebben **en** x **of** y Queer is.
<br><br>

$\forall xy(Bisexueel relatie(x,y) \iff \exists R(x,y) \wedge B(x) \wedge (V(y) \vee M(y) \vee T(y))$

**Voor alle** x geldt dat x een Bisexuele relatie heeft **als** x een Biseksueel persoon is **en** een relatie heeft met y, die een biseksueel persoon is **of** een vrouw **of** een man **of** een Transgender 

Dit verscheelt van een Queer en Interseksuele relatie omdat hier relaties van een Biseksueel persoon met een Queer of Interseksueel persoon niet mee tellen.
<br><br>

```python
F = nx.DiGraph()
M,V= {'Bart','Levi','John','Anton','Laszlo'},{'Sophie', 'Julia','Emma', 'Iris', 'Nina', 'Amy','Fleur'}
MV, VM={'Nova', 'Esmee', 'Amber', 'Yara', 'Clara'},{'Gijs','Duncan','David','Stijn','Minne'}
I = {'Floris','Lila'}
Q={'James','Elton'}
B={'John', 'Anton','Emma','Iris','Amy'}
F.add_nodes_from(M)
F.add_nodes_from(V)
F.add_nodes_from(MV)
F.add_nodes_from(VM)
F.add_nodes_from(I)
F.add_nodes_from(Q)
F.add_nodes_from(B)
for n in M:
    F.nodes[n]['M']=True
    F.nodes[n]['V']=False
    F.nodes[n]['VM']=False
    F.nodes[n]['MV']=False
    F.nodes[n]['I']=False
    F.nodes[n]['Q']=False
    F.nodes[n]['B']=False
for n in V:
    F.nodes[n]['V']=True
    F.nodes[n]['M']=False 
    F.nodes[n]['VM']=False
    F.nodes[n]['MV']=False
    F.nodes[n]['I']=False
    F.nodes[n]['Q']=False
    F.nodes[n]['B']=False
for n in VM:
    F.nodes[n]['V']=False
    F.nodes[n]['M']=False 
    F.nodes[n]['VM']=True
    F.nodes[n]['MV']=False 
    F.nodes[n]['I']=False
    F.nodes[n]['Q']=False
    F.nodes[n]['B']=False
for n in MV:
    F.nodes[n]['V']=False
    F.nodes[n]['M']=False 
    F.nodes[n]['VM']=False
    F.nodes[n]['MV']=True
    F.nodes[n]['I']=False
    F.nodes[n]['Q']=False
    F.nodes[n]['B']=False
for n in I:
    F.nodes[n]['V']=False
    F.nodes[n]['M']=False 
    F.nodes[n]['VM']=False
    F.nodes[n]['MV']=False
    F.nodes[n]['I']=True
    F.nodes[n]['Q']=False
    F.nodes[n]['B']=False
for n in Q:
    F.nodes[n]['V']=False
    F.nodes[n]['M']=False 
    F.nodes[n]['VM']=False
    F.nodes[n]['MV']=False
    F.nodes[n]['I']=False
    F.nodes[n]['Q']=True
    F.nodes[n]['B']=False
for n in B:
    F.nodes[n]['B']=True
    F.nodes[n]['I']=False
    F.nodes[n]['Q']=False
    F.nodes[n]['V']=False or True
    F.nodes[n]['M']=False or True
    F.nodes[n]['VM']=False or True
    F.nodes[n]['MV']=False or True



rel = [('Sophie','Julia'),('Julia','Sophie'),('Bart','Levi'),('Levi','Bart'),
      ('Gijs','Laszlo'),('Laszlo','Gijs'),('Duncan','David'),('David','Duncan'),
      ('Nova','Nina'),('Nina','Nova'),('Yara','Clara'),('Clara','Yara'),('John','Fleur'),('Fleur','John'),
      ('John','Anton'),('Anton','John'),('Emma','Iris'),('Iris','Emma'),('Amy','John'),('John','Amy')
      ,('Esmee','Amber'),('Amber','Esmee'),('Stijn','Minne'),('Minne','Stijn'),('James','Elton'),('Elton','James')
      ,('Lila','Floris'),('Floris','Lila')]
F.add_edges_from(rel)

G=F
def M(x, network=G):
    return network.nodes[x]['M']
def V(x, network=G):
    return network.nodes[x]['V']
def T(x, network=G):
    return network.nodes[x]['VM'] or network.nodes[x]['MV']
def I(x, network=G):
    return network.nodes[x]['I']
def G(x,y, network=G):
    return network.has_edge(x,y) and network.has_edge(y,x)
def Q(x, network=G):
    return network.nodes[x]['Q']
def VM(x, network=G):
    return network.nodes[x]['VM']
def MV(x, network=G):
    return network.nodes[x]['MV']
def B(x, network=G):
    return network.nodes[x]['B']

Man = {x for x in F if M(x,F)}

Vrouw = {x for x in F if V(x,F)}

Transgender = {x for x in F if T(x,F)}

Intersexual = {x for x in F if I(x,F)}

Intersexual_relatie = {(x,y) for x in F for y in F if I(x,F) and I(y,F) and G(x,y,F)}

Homo = {(x) for x in F for y in F if not B(x,F) and
        ((M(x,F) or VM(x,F)) and (M(y,F) or VM(y,F) or (B(y,F) and (M(y,F))) and G(x,y,F)))}

Homo_relatie = {(x,y) for x in F for y in F if not B(x,F) and
        ((M(x,F) or VM(x,F)) and (M(y,F) or VM(y,F) or (B(y,F)and (M(y,F)))) and G(x,y,F))}

Lesbian = {(x) for x in F for y in F if not B(x,F) and 
           (V(x,F)or MV(x,F)) and (V(y,F) or MV(y,F) or (B(y,F) and (V(y,F)))) and G(x,y,F)}

Lesbian_relatie = {(x,y) for x in F for y in F if not B(x,F) and 
           (V(x,F)or MV(x,F)) and (V(y,F) or MV(y,F) or (B(y,F)and (V(y,F)))) and G(x,y,F)}

Queer = {x for x in F if Q(x,F)}

Queer_relatie = {(x,y) for x in F for y in F if Q(x,F) and Q(y,F) and G(x,y,F)}

Bisexuel = {x for x in F if B(x,F)}

Bisexuele_relatie = {(x,y) for x in F for y in F if B(x,F) and G(x,y,F)}



print("Man",Man,'\n')
print("Vrouw",Vrouw,'\n')
print("Transgender",Transgender,'\n')
print("Intersexual",Intersexual,'\n')
print("Homo",Homo,'\n')
print("Homo relatie",Homo_relatie,'\n')
print("Lesbian",Lesbian,'\n')
print("Lesbian relatie",Lesbian_relatie,'\n')
print("Bisexuel",Bisexuel,'\n')
print("Bisexuele relatie",Bisexuele_relatie,'\n')
print("Queer",Queer,'\n')
print("Queer relatie",Queer_relatie,'\n')
print("Intersexual",Intersexual,'\n')
print("Intersexual relatie",Intersexual_relatie,'\n')
```
