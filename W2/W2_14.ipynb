## 1 De gebruikelijke familierelaties

* We werken met predicaatlogische modellen, en met een basistaal met 2 eigenschappen: $M$ en $V$, die staan voor de 2 traditionele geslachten, en 1 twee plaatsige relatie $R$, die staat voor *is ouder van*. Hieronder staat dus dat *piet een man is, en joke een vrouw en dat piet de ouder is van joke*:

$$M(piet), V(joke), R(piet,joke).$$
* Wij gaan nu 2 dingen doen:
    1. beperkingen opleggen aan de relaties
    2. nieuwe relaties definieren in termen van de 3 primtieve relaties $M,V,R$
        * in die nieuwe relaties kan je natuurlijk andere eerder gedefinieerde relaties weer gebruiken.
* **Voorbeelden**
* *Beperking:* 
    * Iedereen is man of vrouw (en nooit allebei tegelijk).
    * Iedereen heeft exact 2 ouders, een man en een vrouw.
* *Nieuwe relatie*
    * $Vader(x,y)$ precies dan en slechts dan alleen als x een ouder is van y en x een man is.
        * $\forall xy( Vader(x,y) \leftrightarrow R(x,y) \wedge M(x))$
    * $Opa(x,y)$ hiervoor gebruik je natuurlijk het $\exists z$    connectief.
        * Je kunt dit netjes definieren in termen van ouder en vader.

*Uitleg PRIMITIEVEN:* <br>
R(x,y): 'x' zou kunnen vallen op 'y', oftewel 'x' zou zich seksueel aangetrokken kunnen voelen door 'y'. <br>
M(x): Man <br>
V(x): Vrouw <br>
T(x): voelt zich van binnen niet als zijn of haar geboorte geslacht <br>
Q(x): iedereen die niet een specifieke seksuele of gender oriëntatie of voorkeur aanhangt of voelt. <br>

<H3>LHBTQI+</H3>

**Lesbienne** <br> 
Vrouw valt alleen op vrouw -->
$\forall xy(Lesbienne(x,y) \leftrightarrow R(x,y) \wedge (V(x)\wedge V(y)))$ <br>


**Homo**<br>
Man valt alleen op man -->
$\forall xy(Homo(x,y) \leftrightarrow R(x,y) \wedge M(x) \wedge M(y))$ <br>

**Biseksueel**<br>
Man of vrouw valt op zowel man als vrouw -->
$\forall xy(Biseksueel(x,y) \leftrightarrow (R(x,y) \wedge M(y)) \wedge (R(x,z) \wedge V(z)))$

**Transgender**<br>
Geboren als geslacht man of vrouw, maar voelt zich niet als aangeboren geslacht -->
$\forall x(Transgender(x) \leftrightarrow T(x))$ <br>

**Queer**<br>
Iedereen die niet een specifieke seksuele of gender oriëntatie of voorkeur aanhangt of voelt --> $\forall x(Queer(x) \leftrightarrow Q(x))$ <br>

**Intersexueel**<br>
Intersekse. Dit betekent dat je zowel aangeboren mannelijke als vrouwelijke kenmerken vertoont. Dit gaat om de aangeboren biologische kenmerken. -->
$\forall x(Interseksueel(x) \leftrightarrow (M(x) \wedge V(x)))$

**+ (plus)**<br>
Staat voor al het onbenoemde wat niet hetero en geen man en geen vrouw. --> 
$\forall xy(plus(x) \leftrightarrow \neg Lesbienne(x) \wedge \neg Homo(x) \wedge \neg Biseksueel(x) \wedge \neg Transgender(x) \wedge \neg Queer(x) \wedge \neg Intersexueel(x) \wedge \neg M(x) \wedge \neg V(x) )$
