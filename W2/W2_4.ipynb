$R(x,y)$ betekent vanaf dit punt "heeft sexuele voorkeur voor"

Extra primitieven:
  - $Onzijdig(x)$ precies dan en slechts dan als x geen man en geen vrouw is.
    - $\forall x( Onzijdig(x) \leftrightarrow \neg M(x) \wedge \neg V(x))$

Voorkeur:
  - $Hetero(x,y)$ precies dan en slechts dan alleen als x een relatie heeft met y, waarbij x en y niet van hetzelfde geslacht zijn. 
    - $\forall xy( Hetero(x,y) \leftrightarrow R(x,y) \wedge (M(x) \wedge \neg M(y)) \wedge (V(x) \wedge \neg V(y) \wedge (O(x) \wedge \neg O(y))$
  - $Homofiel(x,y)$ precies dan en slechts dan alleen als x een relatie heeft met y, waarbij x en y beide mannen zijn. 
    - $\forall xy( Homofiel(x,y) \leftrightarrow R(x,y) \wedge M(x) \wedge M(y))$
  - $Lesbisch(x,y)$ precies dan en slechts dan alleen als x een relatie heeft met y, waarbij x en y beide vrouwen zijn.
