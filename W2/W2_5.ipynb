## Familierelaties
### Ouder
$R(x,y)$: $x$ is de ouder van $y$

### Vader
Iemands vader is diens mannelijke ouder.<br>

$x$ is de vader van $y$ precies dan en slechts dan alleen als $x$ de ouder is van $y$ **EN** $x$ mannelijk is.<br>

$\forall xy(Vader(x,y) \leftrightarrow R(x,y) \wedge M(x))$

### Moeder
Iemands moeder is diens vrouwelijke ouder.<br>

$x$ is de moeder van $y$ precies dan en slechts dan alleen als $x$ de ouder is van $y$ **EN** $x$ vrouwelijk is.<br>

$\forall xy(Moeder(x,y) \leftrightarrow R(x,y) \wedge V(x))$

### Zoon
Iemands zoon is diens mannelijke kind.<br>

$x$ is de zoon van $y$ precies dan en slechts dan alleen als $y$ de ouder is van $x$ **EN** $x$ mannelijk is.<br>

$\forall xy(Moeder(x,y) \leftrightarrow R(y,x) \wedge M(x))$

### Dochter
Iemands dochter is diens vrouwelijke kind.<br>

$x$ is de dochter van $y$ precies dan en slechts dan alleen als $y$ de ouder is van $x$ **EN** $x$ vrouwelijk is.<br>

$\forall xy(Moeder(x,y) \leftrightarrow R(y,x) \wedge V(x))$

### Broer of zus
Iemands broer of zus is iemand anders met minstens Ã©Ã©n dezelfde ouder.

$x$ is de broer of zus van $y$ precies dan en slechts dan alleen als er een $z$ bestaat waarvoor geldt dat $z$ de ouder is van $x$ **EN** $z$ de ouder is van $y$, **EN** $x$ en $y$ niet dezelfde persoon zijn.<br>

$\forall xy(Broerofzus(x,y) \leftrightarrow \exists z(R(z,x) \wedge R(z,y)) \wedge \neg(x = y))$

### Broer
Iemands broer is een ander mannelijk persoon met minstens Ã©Ã©n dezelfde ouder.

$x$ is de broer van $y$ precies dan en slechts dan alleen als er een $z$ bestaat waarvoor geldt dat $z$ de ouder is van $x$ **EN** $z$ de ouder is van van $y$, **EN** $x$ mannelijk is **EN** $x$ en $y$ niet dezelfde persoon zijn.<br>

$\forall xy(Broer(x,y) \leftrightarrow \exists z(R(z,x) \wedge R(z,y)) \wedge M(x) \wedge \neg(x = y))$

### Zus
Iemands zus is een ander vrouwelijk persoon met minstens Ã©Ã©n dezelfde ouder.

$x$ is de zus van $y$ precies dan en slechts dan alleen als er een $z$ bestaat waarvoor geldt dat $z$ de ouder is van $x$ **EN** $z$ de ouder is van van $y$, **EN** $x$ vrouwelijk is **EN** $x$ en $y$ niet dezelfde persoon zijn.<br>

$\forall xy(Zus(x,y) \leftrightarrow \exists z(R(z,x) \wedge R(z,y)) \wedge V(x) \wedge \neg(x = y))$

### Grootouder
De ouders van iemands ouders zijn diegene's grootouders.

$x$ is de grootouder van $y$ precies dan en slechts dan alleen als er een $z$ bestaat waarvoor geldt dat $x$ de ouder is van $z$ **EN** $z$ de ouder is van $y$.<br>

$\forall xy(Grootouder(x,y) \leftrightarrow \exists z(R(x,z) \wedge R(z,y)))$

### Oma
De moeders van iemands ouders zijn diegene's oma's.<br>

$x$ is de oma van $y$ precies dan en slechts dan alleen als er een $z$ bestaat waarvoor geldt dat $x$ de moeder is van $z$ **EN** $z$ de ouder is van $y$.<br>

$\forall xy(Oma(x,y) \leftrightarrow \exists z(Moeder(x,z) \wedge R(z,y)))$

### Opa
De vaders van iemands ouders zijn diegene's opa's.<br>

$x$ is de opa van $y$ precies dan en slechts dan alleen als er een $z$ bestaat waarvoor geldt dat $x$ de vader is van $z$ **EN** $z$ de ouder is van $y$.<br>

$\forall xy(Opa(x,y) \leftrightarrow \exists z(Vader(x,z) \wedge R(z,y)))$

### Oom of tante
De broers en zussen van iemands ouders zijn diegene's ooms of tantes. **N.B.:** De echtgenoten van de broers of zusssen van diens ouders tellen niet mee.

$x$ is de oom of tante van $y$ precies dan en slechts alleen als er een $z$ bestaat waarvoor geldt dat $z$ een ouder is van $y$ **EN** $x$ een broer of zus is van $z$.<br>

$\forall xy(Oomoftante(x,y) \leftrightarrow \exists z(R(z,y) \wedge Broerofzus(x,z)))$

### Oom
Iemands ooms zijn de broers van diegene's ouders.<br>

$x$ is de oom van $y$ precies dan en slechts alleen als $x$ de oom of tante is van $y$ **EN** $x$ mannelijk is.<br>

$\forall xy(Oom(x,y) \leftrightarrow Oomoftante(x,y) \wedge M(x))$

### Tante
Iemands tantes zijn de zussen van diegene's ouders.<br>

$x$ is de tante van $y$ precies dan en slechts alleen als $x$ de oom of tante is van $y$ **EN** $x$ vrouwelijk is.<br>

$\forall xy(Tante(x,y) \leftrightarrow Oomoftante(x,y) \wedge V(x))$

### Nicht 
Je bent iemands nicht als je:
1. dochter bent van diens oom of tante; of
2. dochter bent van diens broer of zus. <br>

$x$ is de nicht van $y$ precies dan en slechts dan alleen als er een $z$ bestaat waarvoor geldt dat 
$x$ de dochter is van $z$ **EN** $z$ de broer of zus is van $y$, **OF** als er een $z$ bestaat waarvoor geldt dat $x$ de dochter is van $z$ **EN** $z$ de oom of tante is van $y$.<br>

$\forall xy(Nicht(x,y) \leftrightarrow \exists z(Dochter(x,z) \wedge Broerofzus(z,y)) \vee \exists z(Dochter(x,z) \wedge Oomoftante(z,y)))$

### Neef
Je bent iemands neef als je:
1. zoon bent van diens oom of tante; of
2. zoon bent van diens broer of zus. <br>

$x$ is de neef van $y$ precies dan en slechts dan alleen als er een $z$ bestaat waarvoor geldt dat 
$x$ de zoon is van $z$ **EN** $z$ de broer of zus is van $y$, **OF** als er een $z$ bestaat waarvoor geldt dat $x$ de zoon is van $z$ **EN** $z$ de oom of tante is van $y$.<br>

$\forall xy(Neef(x,y) \leftrightarrow \exists z(Zoon(x,z) \wedge Broerofzus(z,y)) \vee \exists z(Zoon(x,z) \wedge Oomoftante(z,y)))$

### Achterneef
Wij definiÃ«ren achterneef alleen als een zoon van een neef of nicht en NIET als een zoon van een oudoom of oudtante.

$x$ is de achterneef van $y$ precies dan en slechts dan alleen als er een $z$ bestaat waarvoor geldt dat $x$ de zoon is van $z$ **EN** $z$ de neef is van $y$ **OF** $z$ de nicht is van $y$.<br>

$\forall xy(Achterneef(x,y) \leftrightarrow \exists z(Zoon(x,z) \wedge (Nicht(z,y) \vee Neef(z,y)))$

### Achternicht
Wij definiÃ«ren achternicht alleen als een dochter van een neef of nicht en NIET als een dochter van een oudoom of oudtante.

$x$ is de achternicht van $y$ precies dan en slechts dan alleen als er een $z$ bestaat waarvoor geldt dat $x$ de dochter is van $z$ **EN** $z$ de neef is van $y$ **OF** $z$ de nicht is van $y$.<br>

$\forall xy(Achternicht(x,y) \leftrightarrow \exists z(Dochter(x,z) \wedge (Nicht(z,y) \vee Neef(z,y)))$

## Beperkingen
- Niemand is ouder van zichzelf: $\forall x(\neg R(x,x))$
- Niemand is ouder van zijn of haar eigen ouder: $\forall xy(R(x,y) \rightarrow \neg R(y,x))$
- Iedereen is man of vrouw, en nooit beiden tegelijk: $\forall x(M(x) \veebar V(x))$
  - N.B.: Dit is, uiteraard, omstreden. Dit is echter een noodzakelijke beperking voor deze opdracht om de relaties enigszins redelijk te kunnen definiÃ«ren.
- Niemand is zowel vader als moeder: $\forall x(\exists y(Vader(x,y)) \leftrightarrow \neg \exists y(Moeder(x,y)))$
- Iedereen die een broer of zus heeft is ook de broer of zus van die persoon: $\forall xy(Broerofzus(x,y) \leftrightarrow Broerofzus(y,x))$
- Niemand is de ouder van zijn of haar broer of zus: $\forall xy(R(x,y) \rightarrow \neg Broerofzus(x,y))$

# Uitwerking van opdracht 2

## Primitieven
$M(x)$ betekent dat $x$ zich identificeert als een man <br>
$V(x)$ betekent dat $x$ zich identificeert als een vrouw <br>
$MG(x)$ betekent dat $x$ geboren is met mannelijke geslachtseigenschappen <br>
$VG(x)$ betekent dat $x$ geboren is met vrouwelijke geslachtseigenschappen <br>

$R(x,y)$ betekent dat $x$ zich romantische of seksueel aangetrokken voelt tot $y$

## Letters
### Heteroseksueel
Iemand is heteroseksueel wanneer deze persoon zich aangetrokken voelt tot iemand en zich tot niemand aangetrokken voelt die zich met hetzelfde geslacht identificeert.

$\forall x(Heteroseksueel(x) \leftrightarrow \exists y(R(x,y)) \wedge ((M(x) \wedge \neg \exists y(R(x,y) \wedge M(y))) \vee (V(x) \wedge \neg \exists y(R(x,y) \wedge V(y)))))$

### Homoseksueel
Iemand is homoseksueel wanneer deze persoon zich identificeert als man, hij zich aangetrokken voelt tot minstens Ã©Ã©n persoon en hij zich tot niemand aangetrokken voelt die zich identificeert als vrouw.

$\forall x(Homoseksueel(x) \leftrightarrow M(x) \wedge \exists y(R(x,y)) \wedge \neg \exists y(V(y) \wedge R(x,y)))$

### Lesbisch
Iemand is lesbisch wanneer deze persoon zich identificeert als vrouw, zij zich aangetrokken voelt tot minstens Ã©Ã©n persoon en zij zich tot niemand aangetrokken voelt die zich identificeert als man.

$\forall x(Lesbisch(x) \leftrightarrow V(x) \wedge \exists y(R(x,y)) \wedge \neg \exists y(M(y) \wedge R(x,y)))$

### Biseksueel
Iemand is biseksueel wanneer deze persoon zich aangetrokken voelt tot minstens Ã©Ã©n persoon die zich als man identificeert en Ã©Ã©n persoon die zich als vrouw identificeert. **N.B.:** Hierbij maken wij duidelijk onderscheid tussen biseksualiteit en panseksualiteit, oftewel het vallen op een mens ongeacht de persoon hun sekse of genderidentiteit.<br>

$\forall x(Biseksueel(x) \leftrightarrow \exists y(R(x,y) \wedge M(y)) \wedge \exists y(R(x,y) \wedge V(y)))$

### Transgender
Iemand is transgender wanneer hij zich identificeert als man maar geboren is zonder mannelijke geslachtseigenschappen en met vrouwelijke geslachtseigenschappen, of zij zich identificeert als vrouw maar geboren is zonder vrouwelijke geslachtseigenschappen en met mannelijke geslachtseigenschappen.

$\forall x(Transgender(x) \leftrightarrow (M(x) \wedge \neg MG(x) \wedge VG(x)) \vee (V(x) \wedge \neg VG(x) \wedge MG(x)))$

### Queer / Questioning
Iemand is queer/questioning wanneer deze persoon zich niet identificeert als man en zich ook niet identificeert als vrouw.

$\forall x(Queer(x) \leftrightarrow \neg V(x) \wedge \neg M(x))$

### Interseksueel
Iemand is interseksueel wanneer deze persoon zowel mannelijke als vrouwelijke geslachtseigenschappen heeft.

$\forall x(Interseksueel(x) \leftrightarrow VG(x) \wedge MG(x))$

### +
Wij houden voor de **+** aan dat deze persoon zich niet identificeert met de bovengenoemde termen voor seksuele orientatie en genderidenteit. <br>
$\forall x(Plus(x) \leftrightarrow \neg Heteroseksueel(x) \wedge \neg Homoseksueel(x) \wedge \neg Lesbisch(x) \wedge \neg Biseksueel(x) \wedge \neg Transgender(x) \wedge \neg Queer(x) \wedge \neg Interseksueel(x))$
