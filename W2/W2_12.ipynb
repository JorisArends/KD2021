### 2. 

Een relatie die niet in predikaatlogica gedefinieerd kan worden is de stiefvader relatie. Er kan alleen aangegeven worden of een persoon een ouder is en zijn geslacht. Tussen de stiefzoon en stiefvader is geen ouderrelatie. Daarnaast kan je niet via de moeder bij de stiefvader uitkomen, omdat deze ook geen vorm van ouder relatie heeft met de stiefvader. Om stiefvaders te kunnen definiëren moet er dus een extra relatie aan de basistaal worden toegevoegd in de vorm van een tweezijdige stiefouder relatie of een partnerrelatie.
 
### 3.

Niemand is ouder van zichzelf:

- $ \neg∃x(R(x,x))$


Er is niemand met twee vaders:

- $ \neg∃x∃yz(M(y) \wedge M(z) \rightarrow R(y,x) \wedge R(y,x))$

Er is niemand met twee moeders:

- $ \neg∃x∃yz(V(y) \wedge V(z) \rightarrow R(y,x) \wedge R(y,x))$

Iedereen is man of vrouw:

- $ ∀x((M(x) \vee V(x)) \wedge \neg(M(x) \wedge V(x)) $

Iedereen heeft exact twee ouders:

- $\forall x \exists yz \neg \exists w (R(y,x) \wedge R(z,x) \wedge R(w,x)) $

**Base predicates:**

- MV(x) = voelt zich man; x heeft genderidentiteit man
- VV(x) = voelt zich vrouw; x heeft genderidentiteit vrouw
- OV(x) = voelt zich geen man of vrouw; x heeft genderidentiteit overig
- M(x) = is biologische man; x is biologisch gezien een man
- V(x) = is biologische vrouw; x is biologisch gezien een vrouw
- R(x,y) = valt op; x valt op y

**Met sekse wordt bedoeld het biologische geslacht waarmee een persoon is geboren.**

Persoon waarbij de genderidentiteit overeenkomt met de sekse:

Cisgender: $\forall x ((M(x) \wedge MV(x)) \vee (V(x) \wedge VV(x)))$

Persoon met genderidentiteit vrouw die op genderidentiteit vrouw valt:

Lesbisch: $∀x∃yz(VV(x) \wedge VV(y) \wedge MV(z) \rightarrow R(x,y) \wedge \neg R(x,z))$

Persoon met genderidentiteit man die op genderidentiteit man valt:

Homosexueel: $∀x∃yz(MV(x) \wedge MV(y) \wedge VV(z) \rightarrow R(x,y) ^ \neg R(x,z))$

Persoon die op sekse man en vrouw valt:

Bisexueel: $\forall x \exists y (MV(x) \vee VV(x) \rightarrow R(x,y))$

Persoon waarbij de genderidentiteit niet overeenkomt met de sekse:

Transgender: $\forall x (M(x) \rightarrow VV(x) \vee V(x) \rightarrow MV(x))$

Persoon die geen vaste genderidentiteit en valt op alle genderidentiteiten: 

Queer: $\forall x (OV(x) \wedge Pansexueel(x))$ 

Persoon die op geen enkele genderidentiteit valt:

Asexueel:$ \forall x \neg \exists y (R(x,y))$

Persoon die niet geboren is met sekse man of vrouw:

Interseksueel:  $\forall x \neg M(x) ^ \neg V(x))$

Persoon die op alle genderidentiteiten valt:

Pansexueel: $\forall xy (MV(y) \vee VV(y) \vee OV(y) \rightarrow R(x,y))$

Persoon met sekse man die op sekse vrouw valt of persoon met sekse vrouw die op sekse man valt:

Hetero: $\forall xy (MV(x) \wedge R(x,y) \rightarrow VV(y) \vee VV(x) \wedge R(x,y) \rightarrow MV(y))$

Persoon dat zich met geen van de LHBTQI 'labels' associeert:

Plus: $\forall x \neg (Hetero(x) \vee Pansexueel(x) \vee Intersexueel(x) \vee Asexueel(x) \vee Queer(x) \vee Transgender(x) \vee Bisexueel(x) \vee Homosexueel(x) \vee Lesbisch(x) \vee Cisgender(x))$

Beperkingen:
Plus zit niet in het netwerk hieronder, omdat voor onze definitie van Plus niet een specifiek geval te omschrijven is.

```python
G = nx.DiGraph()

# Maak nodes aan en geef ze hun gender voorkeur
nodes = {
    'Cis':[True,True,False,False,False], # We hebben het cisgender persoon ook hetero gemaakt
    'Lesbisch':[False,False,False,True,True], 
    'Homo':[True,True,False,False,False], 
    'Bi':[False,False,False,True,True], 
    'Trans':[True,False,False,True,False], # Trans persoon is ook hetero 
    'Asexueel':[False,False,False,True,True], 
    'Intersexueel':[False,True,False,False,False], # Intersexueel persoon is ook homo
    'Pansexueel':[False,False,False,True,True], 
    'Queer':[False,False,True,False,True], 
    'Queer2':[True,False,True,False,False]}

G.add_nodes_from(nodes)

# Geef de juiste voorkeur aan elke node
for n,bools in nodes.items():
    G.nodes[n]['M']=bools[0]
    G.nodes[n]['MV']=bools[1]
    G.nodes[n]['OV']=bools[2]
    G.nodes[n]['VV']=bools[3]
    G.nodes[n]['V']=bools[4]

# Alle 'valt op' relaties lekker niet gehardcode
valt_op_vrouwen = ['Cis', 'Lesbisch', 'Bi', 'Pansexueel', 'Queer', 'Queer2']
valt_op_mannen = ['Bi', 'Homo', 'Trans', 'Intersexueel', 'Pansexueel', 'Queer', 'Queer2']
valt_op_overig = ['Queer', 'Queer2', 'Pansexueel']
orientatie = [(valt_op_vrouwen, 'VV'), (valt_op_mannen, 'MV'), (valt_op_overig, 'OV')]

relaties = [(node, node2) for voorkeur,gevoel in orientatie for node in voorkeur for node2 in nodes
            if G.nodes[node2][gevoel] and node != node2]

G.add_edges_from(relaties)
        
# Alle primitieven als functie
def MV(x, G=G):
    return G.nodes[x]['MV']
def VV(x, G=G):
    return G.nodes[x]['VV']
def OV(x, G=G):
    return G.nodes[x]['OV']
def M(x, G=G):
    return G.nodes[x]['M']
def V(x, G=G):
    return G.nodes[x]['V']
def R(x,y,G=G):
    return G.has_edge(x,y)


''' Alle orientaties als comprehensions
    De print statements zijn om te testen '''


# Persoon waarbij de genderidentiteit overeenkomt met de sekse:
Cisgender = {x for x in G if (M(x) and MV(x)) or (V(x) and VV(x))}
print('Cisgender: ', Cisgender)

# Persoon met genderidentiteit vrouw die op genderidentiteit vrouw valt:
Lesbisch = {x for x in G for y in G for z in G if (VV(x) and VV(y) and MV(z) and R(x,y,G) and not R(x,z))}
print('\nLesbisch: ',Lesbisch)

# Persoon met genderidentiteit man die op genderidentiteit man valt:
Homosexueel = {x for x in G for y in G for z in G if (MV(x) and MV(y) and VV(z) and R(x,y,G) and not R(x,z))}
print('\nHomo: ',Homosexueel)

# Persoon die op genderidentiteit man en vrouw valt:
Bisexueel = {x for x in G if {y for y in G if MV(y) and R(x,y)} and {y for y in G if VV(y) and R(x,y)} and not {y for y in G if OV(y) and R(x,y)}}
print('\nBisexueel: ',Bisexueel)

# Persoon waarbij de genderidentiteit niet overeenkomt met de sekse:
Transgender = {x for x in G if (M(x) and VV(x)) or (V(x) and MV(x))}
print('\nTransgender: ',Transgender)

#Persoon die op geen enkele genderidentiteit valt:
Asexueel = {x for x in G if not {y for y in G if R(x,y)}}
print('\nAsexueel: ',Asexueel)

#Persoon die niet geboren is met sekse man of vrouw:
Intersexueel = {x for x in G if not M(x) and not V(x)}
print('\nIntersexueel: ',Intersexueel)

# Pansexueel functie wordt gebruikt in pansexuelen en queer definitie
def Pansexueel(x,G=G):
    if {y for y in G if MV(y) and R(x,y)} and {y for y in G if VV(y) and R(x,y)} and {y for y in G if OV(y) and R(x,y)}:
        return x

# Persoon die op alle genderidentiteiten valt:
Pansexuelen = {x for x in G if Pansexueel(x)}
print('\nPansexueel: ',Pansexuelen)

# Persoon met sekse man die op sekse vrouw valt of persoon met sekse vrouw die op sekse man valt: 
Hetero = {x for x in G for y in G for z in G if x != z and ((MV(x) and R(x,y) and VV(y) and MV(z) and not R(x,z)) or (VV(x) and R(x,y) and MV(y) and VV(z) and not R(x,z)))}
print('\nHetero: ',Hetero)

# Persoon die geen vaste genderidentiteit heeft en valt op alle genderidentiteiten: 
Queer = {x for x in G if OV(x) and Pansexueel(x)}
print('\nQueer: ',Queer)

# None of the above
Plus = {x for x in G if not x in Cisgender | Lesbisch | Homosexueel | Bisexueel | Transgender | Asexueel | Intersexueel | Pansexuelen | Hetero | Queer}
print('\nPlus: ', Plus)

```
