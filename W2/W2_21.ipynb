* $Bisexueel = \forall x \exists y(B(x) \leftrightarrow R(x,y) \wedge (M(x) \vee  V(x)) \wedge (M(y) \vee V(y))$
* $Transgender = \forall x(T(x) \leftrightarrow (M(x) \wedge BV(x)) \vee (V(x) \wedge BM(x))$
